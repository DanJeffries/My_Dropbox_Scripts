```python
pops_list = []
pops_file = open('../Cstacks/crucian_only_catalog_M8_N8/pop_codes_final.txt', 'r')
PCS_file = open('../populations/crucian_only_M8_N8/r_0.7_P_17/PCS_with_names.txt' ,'r')
for line in pops_file.readlines():
    pops_list.append(line.split()[1])
pops_list = set(pops_list)
print pops_list

for pop in pops_list:
    for line in open('../populations/crucian_only_M8_N8/r_0.7_P_17/PCS_with_names.txt' ,'r').readlines():
        if pop in line:
            f = open('../populations/crucian_only_M8_N8/r_0.7_P_17/PCS_'+pop+'.txt', 'a')
            f.write(line+'\n')
            f.close()
            print line
    print '###########################################'
```
