### The sub-samples need to have
- varying numbers of bottlenecked and non-bottlenecked pops in them
- no two subsamples can be the same



```python
%matplotlib inline
import random
import matplotlib.pyplot as plt
from collections import Counter

### Need to ammend the below script to produce an even distribution of the number of bottlenecked and non-bottlenecked pops
### Probs just easier to do this post hoc.

root = !pwd
print root, "\n"

def random_maker(number_of_lists):
    
    ## First, make a list of 7 random from 0-18 (1-19 in non python terms).

    hist_data = []
    random_dict = {}
    for the_list in range(1,(number_of_lists+1),1):
        random_dict[the_list] = []
  
        ## Make the dictionary
        
        bottled = 0
        for i in range (100): ## iterate process 100 times to ensure I'm going to fill the list with uniq values. 
            if len(random_dict[the_list]) < 7:
                new_rand = (random.randrange(1,19,1))
                if new_rand not in random_dict[the_list]:
                    random_dict[the_list].append(new_rand)
                if new_rand < 10:
                    bottled +=1
        hist_data.append(bottled)   
    ## Make sure the dictionary doesn't have repeated sequences in it?? Does this take the "random" out of things...?
    
    print "length of random = ", len(random_dict)
    
        
    # Extract the dictionary into a list of (key, value) tuples.
    tuples =  [(key, sorted(random_dict[key])) for key in random_dict]

    # Sort the list -- by default it will sort by the key since it is first in the tuple.
    
    tuples.sort()

    # Reset the dictionary so it is ready to hold the new dataset.
    random_dict_set = {}

    # Load key-values into the dictionary. Only the first value will be
    # stored.
    for k, v in tuples:
        if sorted(v) in random_dict_set.values():
            continue
        random_dict_set[k] = sorted(v)

    print "length of setted random = ", len(random_dict_set)
        
   
    ## Make a quick hist of the number of uniq sub-samples in each class
    plt.hist(hist_data)
    plt.title("Number of Bottlenecked pops in sub-sample")
    plt.axes()
    plt.show()
    
   # print random_dict
    return random_dict


sub_samples= random_maker(100)
```
