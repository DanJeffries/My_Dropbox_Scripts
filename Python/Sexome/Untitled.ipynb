```python
import sys

filepath = sys.argv[1]

tagsfile = open(filepath, 'r').readlines()[1:]

covs = open("%s_covs.text" % filepath.rpartition(".")[0])

count = 0
for line in tagsfile:
    #print len(line.split())
    if "consensus" in line.split():
        covs.write("%s\n" % count)
        count = 0
        
    elif "primary" in line.split() or "secondary" in line.split():
        count += 1
```

```python
from matplotlib import pyplot as plt
import numpy as np
print np.mean(covs)
plt.hist(covs, bins=10000)
plt.xlim(0,50)
plt.show()
```
