## So first, get the highest quality mappings from the Blast mapping file


```python
###### Getting gene models from BLAST mapping ######

## First, because in the genome file, each scaff has two names, their index and the ENA ID
## I need to get corresponding ID's. The "Scaffold_names.txt" file below is simlpy the headers 
## for each scaffold grepped from the genome fasta. 

Local_draft_Scaffolds = open("/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Common_carp_genome/Xu_et_al_2014_Nature_version/Scaffold_names.txt", 'r').readlines()

## Next load the mapping locations for my mapping of BAYENV winners to the XU et al genome "PRJEB7241_all.fasta"
BAYENV_winners_raw = open("/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/BAYENV/Sub-samples/MASTER_npops_17/MASTER_winners_blast_results.out", 'r')

## Finally get their gff file, which contains the locations of the transcripts that they have mapped onto the genome.
#Gff_scaffolds = open("./V1_gff_scaffs.txt", 'r').readlines()

gff = open("/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Common_carp_genome/Xu_et_al_2014_Nature_version/genemodels/V1.0.gff").readlines()

### Using the above, I can retrieve the ID's of the nearest transcript to the locations that my BAYENV winners mapped to


## In blast output - Column 4 contains the length of the alignment
## my sequences are 92 bp, so I want everything that is has an alignment length of above 89 (got this from a quick histogram)

## I want to keep the Stacks locus ID, the scaffold ID and the mapping coordinate

Bayenv_winners_IDs = [] ## contains tuples of locus ID, the scaffold ID and the mapping start coordinate

for line in BAYENV_winners_raw:
    if int(line.split()[3]) >= 85:
        Bayenv_winners_IDs.append((line.split()[0], line.split()[1].split('|')[1], line.split()[8]))

print "Number of High quality alignments = ", len(Bayenv_winners_IDs)

        
print "SNP locus\t", "Scaffold\t", "map_start_coord"
for winner in Bayenv_winners_IDs:
    print winner
    


### Note - all e values are extremely low, so I have only used an alignment length cut off here
```

## Then get the other ID for each scaffold, ie, the ID Xu et al used in their transcriptome mapping

```python
Bayenv_winner_scaffs = [("STACKS_ID", "XU_scaff_ID", "Map_start_coord")]

for a,b,c in Bayenv_winners_IDs:
    for scaff in Local_draft_Scaffolds:
        if b == scaff.split("|")[1]:
            Bayenv_winner_scaffs.append((a, scaff.split()[-1], c))
print "Bayenv_winner_scaffs"
for winner in Bayenv_winner_scaffs:
    print winner

## so this tuple contains [0] my RAD locus ID, [1] XU et al scaff names, [0] location that my BAYENV winners map on each scaffold
```

## Finally:
- Get the transcripts from each scaffold, within a set distance from where my SNP locus maps
- Output this, along with all of the information above into a file

```python
Bayenv_GMs = open("/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/BAYENV/Sub-samples/MASTER_npops_17/MASTER_winners_genemodels_3kb_LBs_85bp_alignment_length.txt", 'w')

Bayenv_GMs.write("##STACKS_ID\tSTACKS_loc_mapping_Start\tXU_scaff_ID\tTranscript_name\tTranscript_mapping_start\tSwiss-Prot_function\n")  ## Headers
for a,b,c in Bayenv_winner_scaffs[1:]:
    for line in gff:
        if b == line.split()[0] and (int(c)-3000) <= int(line.split()[3]) <= (int(c)+3000): ## this is where I am controlling the size of my linkage blocks . . .
            ## Get the GO functions from the Swissprot output
            swissprot = open("/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Common_carp_genome/Xu_et_al_2014_Nature_version/genemodels/Commoncarp_swissprot.alignment", 'r').readlines()
            for swissline in swissprot:
                if line.split()[8] == swissline.split()[0]:
                    swiss_func = swissline.split()[11:]
            Bayenv_GMs.write(a+"\t"+str(c)+"\t"+line.split()[0]+"\t"+line.split()[8]+"\t"+str(line.split()[3])+"\t"+" ".join(swiss_func)+"\n")
Bayenv_GMs.close()
```
