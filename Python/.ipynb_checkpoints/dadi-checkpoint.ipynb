```python
### Note that this function NOW works for vcf files containing many pops!

def vcf_to_dadi(VCF_file, outgrp_list, out_file, blacklist=[]):
    
    VCF_handle = open(VCF_file, 'r') ## make handle
    
    my_vcf = vcf.Reader(VCF_handle) ## read in data with vcf module
    
    ## Get blacklist if its there...
    my_blacklist = []
    
    if blacklist:
        my_blacklist = open(blacklist, 'r').readlines()
        my_blacklist = [i.strip() for i in my_blacklist]
        
    ### Get population names (from first snp in file), and write header line to input file ###
    
    snp = my_vcf.next()
    names = []
    for sample in snp.samples:
        names.append(sample.sample.split("_")[0].strip("0123456789-"))
    names = sorted(set(names))
    print "Populations (n="+str(len(names))+"):" 
    print ', '.join(names)
    header_line = "In_REF\tOut_REF\tAllele1\t"+'\t'.join(names)+"\tAllele2\t"+'\t'.join(names)+"\tTag_ID\tSite"
    
    dadi_input = open(out_file, 'w')
    
    dadi_input.write(header_line+"\n") ## write header line
    
    
    ##### Now get the data ######
    
    VCF_handle = open(VCF_file, 'r') ## make handle
    my_vcf = vcf.Reader(VCF_handle) ## Parse file again, as .next() will have moved the iterator on
    
    ## For each SNP in the file ...
    
    total_snp_counter = 0
    blacklist_snp_counter = 0
    missing_in_outgrp_snp_counter = 0
    
    ## debugging counters
    
    high_out_alt_counter = 0
    high_out_ref_counter = 0
   
    high_in_alt_counter = 0
    high_in_ref_counter = 0
    
    for snp in my_vcf:
        total_snp_counter += 1
        if not snp.ID in my_blacklist :
            blacklist_snp_counter += 1
            
            out_goup_genotypes = (snp.genotype(i)['GT'] for i in outgrp_list)
                           
            GT_dict = {}                
            ## get outgroup ref allele and data                
            out_ref_count = 0
            out_alt_count = 0
                
            for sample in snp.samples:
                if any([sample.sample.startswith(a) for a in outgrp_list]):
                    if sample['GT'] == "0/0":
                        out_ref_count += 2
                    elif sample['GT'] == "0/1":
                        out_ref_count += 1
                        out_alt_count += 1  ## counting one allele twice
                    elif sample['GT'] == "1/0":
                        out_ref_count += 1
                        out_alt_count += 1
                    elif sample['GT'] == "1/1":
                        out_alt_count += 2
                
            if all("./." == gt for gt in out_goup_genotypes):
                OUT_REF_ALLELE = snp.REF
                missing_in_outgrp_snp_counter += 1
            elif out_ref_count < out_alt_count: ## changed this from <= to < so all heterozygous snps in RM31, 32 will be assigned the ingroup major allele - this has drastically changed the input file, the outgroup ref more commonly 
                OUT_REF_ALLELE = snp.ALT[0]
                GT_dict["Outgroup"]=(out_alt_count,out_ref_count)
                high_out_alt_counter += 1
            elif out_alt_count <= out_ref_count:
                OUT_REF_ALLELE = snp.REF
                high_out_ref_counter +=1
                GT_dict["Outgroup"]=(out_ref_count,out_alt_count) 
                
                
                #### for all non outgroup populations ####
        
        
            for population in names:
                in_ref_count = 0
                in_alt_count = 0             
                GT_dict[population] = ()        
                for sample in snp.samples:
                    if sample.sample.startswith(population) and not any([sample.sample.startswith(a) for a in outgrp_list]):
                        if sample['GT'] == "0/0":
                            in_ref_count += 2
                        elif sample['GT'] == "0/1":
                            in_ref_count += 1
                            in_alt_count += 1
                        elif sample['GT'] == "1/0":
                            in_ref_count += 1
                            in_alt_count += 1
                        elif sample['GT'] == "1/1":
                            in_alt_count += 2                            
                    
                    #print population, in_ref_count, in_alt_count
                    
                if in_alt_count <= in_ref_count: ## incase the common carp genotype pushes the major allele counts the wrong way
                    IN_REF_ALLELE = snp.REF
                    IN_ALT_ALLELE = snp.ALT[0]
                    GT_dict[population]=(in_ref_count,in_alt_count)
                    high_in_ref_counter += 1
                elif in_ref_count < in_alt_count:
                    IN_REF_ALLELE = snp.ALT[0]
                    IN_ALT_ALLELE = snp.REF
                    GT_dict[population]=(in_alt_count, in_ref_count)
                    high_in_alt_counter += 1
                                    
            in_REF_counts =[]
            in_ALT_counts =[]
                
            for population in names:
                in_REF_counts.append(str(GT_dict[population][0]))
                in_ALT_counts.append(str(GT_dict[population][1]))
        
        
            snp_line = "-"+str(IN_REF_ALLELE)+"-"+"\t"+"-"+str(OUT_REF_ALLELE)+"-"+"\t"+str(IN_REF_ALLELE)+"\t"+str('\t'.join(in_REF_counts))+"\t"+str(IN_ALT_ALLELE)+"\t"+str("\t".join(in_ALT_counts))+"\t"+str(snp.ID)+"\t"+str(snp.POS) 
            dadi_input.write(snp_line+"\n")
    print "Number of blacklisted snps ignored =", total_snp_counter - blacklist_snp_counter       
    print "Number of SNPs processed = ", total_snp_counter
    print "Number of snps missing in outgroup (ingroup ref assigned) =", missing_in_outgrp_snp_counter
    print "Number of snps used and written to file =", total_snp_counter
    print "Number of snps where outgroup ref count higher than alt count =",  high_out_ref_counter
    print "Number of snps where outgroup alt count higher than ref count =",  high_out_alt_counter
        
    print "Number of snps where ingroup alt count higher than ref count = roughly", high_in_alt_counter/len(names)
    print "Number of snps where ingroup ref count higher than ref count = roughly", high_in_ref_counter/len(names)
```

### Defining a model

```python
help(dadi.Numerics.default_grid)
```

### Parameters =
- nuB = depth of bottleneck
- nuF recovered to this size
- TF= time into the past that the population recovered to nuF
- TB = time into the past that the bottleneck nuB occurred

### Units
- Times are given as 2Nref i.e 2x the N of the reference population.
- nuB and nuF are relative to the reference population I think . . . 

```python
def bottleneck(params, ns, pts):
    nuB, nuF, TB, TF = params 
    xx = dadi.Numerics.default_grid(pts)
    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.Integration.one_pop(phi,xx,TB,nuB)
    phi = dadi.Integration.one_pop(phi,xx,TF,nuF)
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,))
    return fs
```

```python
def multi_integration(params, ns, pts):
    nuA, nuB, nuC, TA, TB, TC = params 
    xx = dadi.Numerics.default_grid(pts)
    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.Integration.one_pop(phi,xx,TA,nuA)
    phi = dadi.Integration.one_pop(phi,xx,TB,nuB)
    phi = dadi.Integration.one_pop(phi,xx,TC,nuC)
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,))
    return fs
```

### Specify starting values for the model

```python
## For bottleneck model recent decline

NB = 0.1 ## bottlenecked size
NF = 1 ## recovered size
TiB = 2 ## time of bottleneck in the past, since recovery (i.e. TiF + TiB is total time since bottleneck)
TiF = 4 ## time in the past since recovery

parameters = (NB, NF, TiB, TiF)
nsampls = [FS.sample_sizes]
points_l = 15 ## Points to extrapolate

## search limits
Upper = [1, 1, 5, 10]
Lower = [1e-3, 1e-3, 1e-1, 1e-1]

## Make extrapolation function
#bottleneck_extrap = dadi.Numerics.make_extrap_log_func(bottleneck)

## Perturb start values
p0 = dadi.Misc.perturb_params(parameters, fold = 1, upper_bound=Upper, lower_bound=Lower)
```

### Optimise parameters for the model

```python
## Optimise from the starting position for the best params to use
Maxiters = 10

pOpt = dadi.Inference.optimize_log_lbfgsb(p0, FS, bottleneck, points_l, Lower, Upper, maxiter=Maxiters)

print "model parameters =", zip(["NB", "NF", "TiB", "TiF"], pOpt)
```

### Run Model and calculate log-likelihood 

```python

botl_model = bottleneck(pOpt, nsampls, points_l)
ll_botl_model = dadi.Inference.ll_multinom(botl_model, FS)

print "Log-likelihood of model =", ll_botl_model
```
