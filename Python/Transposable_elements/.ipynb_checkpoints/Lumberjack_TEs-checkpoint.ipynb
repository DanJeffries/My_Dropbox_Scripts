### Define functions

A customised version of DansBlastParse that outputs sequences of TE hits

```python
def DansBlastParse_custom(blast_xml_output, best_hit_criteria, Eval_thresh):
    """
    
    Usage: BlastParseExtra.py  <blast_xml_output>  <best_hit_criteria>  <Evalue_threshold>
    
    <blast_xml_output>  -  Path should be absolute
    <best_hit_criteria> -  Orders of magnitude higher that best hit has to be in multi
    <Evalue_threshold>  -  (1e-20 is default)
    
    
    This script first filters the mappings in the <blast_xml_output>, for only those where the
    best hit is at least 5 orders of magnitude higher than the second.
    
    It will then retrieve a segment of the <genome_fasta> which is + and - the <window_size> around the 
    mapping coordinates for each query. If the ends of the scaffold are not within this window, then the 
    length of the segment will be (length of mapped query sequence + 2 x <window_size>). However if an end
    of a scaffold is within this window, the segment will be trimmed to this length.
    
    """

    blast = NCBIXML.parse(open(blast_xml_output,"r"))

    good_blast_counts = 0
    uniq_counts = 0
    
    blast_hit_IDs = []
    
    
    for record in blast :
        #print len(record.alignments)
        if len(record.alignments)==1:  ## Extra else statement - this wasn't in Alan's original script.
            if float(record.alignments[0].hsps[0].expect) < Eval_thresh:
                subject = str(record.alignments[0].hit_def)
                Evalue = float(record.alignments[0].hsps[0].expect)
                Hit_start_coord = int(record.alignments[0].hsps[0].sbjct_start)
                Hit_end_coord = int(record.alignments[0].hsps[0].sbjct_end)
                
                good_blast_counts += 1
                uniq_counts += 1
                blast_hit_IDs.append(subject) ### get the subject ID for the hit!
                
                
                
        elif len(record.alignments)>1:
            if all([record.alignments[0].hsps[0].expect < best_hit_criteria * record.alignments[1].hsps[0].expect, record.alignments[0].hsps[0].expect <= Eval_thresh]):
                
                subject = str(record.alignments[0].hit_def)
                Evalue = float(record.alignments[0].hsps[0].expect)
                Hit_start_coord = int(record.alignments[0].hsps[0].sbjct_start)
                Hit_end_coord = int(record.alignments[0].hsps[0].sbjct_end)
                
                good_blast_counts += 1
                blast_hit_IDs.append(subject) ### get the subject ID for the hit!
                

    
    return good_blast_counts, blast_hit_IDs
```

A function to find lumberjack (LJ) stacks from a catalog.tags.tsv file from Stacks. Returns the IDs of LJ loci.

```python
def Get_lumberjacks(tags_file_path):
    
    import gzip
    
    if tags_file_path.endswith("gz"):
        tags_file = gzip.open(tags_file_path, 'r').readlines()
    else:
        tags_file = open(tags_file_path, 'r').readlines()
        
    Blacklisted_loci = []

    ## First pass - get the locus IDs of the blacklisted loci


    for line in tags_file:

        if "consensus" in line.split():


            Tag_ID = line.split()[2]

            ## Check to see if the locus was a lumberjack or not

            if line.split()[9] == "1" and line.split()[8] == "1" :
                Blacklisted_loci.append(Tag_ID)
    
    print "Number of lumberjacks = %s" % len(Blacklisted_loci)
    
    return Blacklisted_loci

                
```

A function to find the coverages of the LJs identified by Get_lumberjacks()

```python
def Get_lumber_covs(tags_file_path, lumberjack_loci):
    
    import gzip
    from collections import Counter
        
    lumber_covs = {}
    
    if tags_file_path.endswith("gz"):
        tags_file = gzip.open(tags_file_path, 'r').readlines()
    else:
        tags_file = open(tags_file_path, 'r').readlines()
    
    read_tag_IDs = []
    
    for line in tags_file:
        
        read_tag_ID = line.split()[2]
        read_tag_IDs.append(read_tag_ID)
        
    c = Counter(read_tag_IDs)
    
    for lumberjack in lumberjack_loci:
        if lumberjack in read_tag_IDs:
            lumber_covs[lumberjack] = c[lumberjack]-2 ## -2 to remove the consensus and model lines for each locus
        else:
            print lumberjack, "not in read_tag_IDs"
        
        
    return lumber_covs
        
```

A function to find the average coverage of non-lumberjack (and non blacklisted) loci in a sample

```python
def Get_NON_lumber_covs(tags_file_path):
    
    import gzip
    from collections import Counter
        
    Non_lumber_covs = []
    
    if tags_file_path.endswith("gz"):
        tags_file = gzip.open(tags_file_path, 'r').readlines()
    else:
        tags_file = open(tags_file_path, 'r').readlines()
    
    Good_tags = []
    
    ## 1st pass to find non_blacklisted markers
    
    for line in tags_file:
        if all(["consensus" in line, line.split()[8] == "0", line.split()[9] == "0"]):
            Good_tags.append(line.split()[2])
            
    ## 2nd pass to get counts
    
    read_tag_IDs = []
    
    for line in tags_file[1:]: ## not title line
        read_tag_ID = line.split()[2]
        read_tag_IDs.append(read_tag_ID)

    c = Counter(read_tag_IDs)
    
    for ID in c:
        if ID in Good_tags:
            Non_lumber_covs.append(c[ID]-2) ## -2 to remove the consensus and model lines for each locus
            
    print "Average non-lumber coverage = %s" % np.mean(Non_lumber_covs)
    
    return Non_lumber_covs
```

A function to get the sequences for the LJs from a catalog

```python
def Get_lumber_seqs(tags_file_path, lumberjack_loci):
    
    import gzip
    from collections import Counter
        
    lumber_seqs = {}
    
    if tags_file_path.endswith("gz"):
        tags_file = gzip.open(tags_file_path, 'r').readlines()
    else:
        tags_file = open(tags_file_path, 'r').readlines()
    
    for line in tags_file:
        if "consensus" in line:
            tag_ID = line.split()[2]
            if tag_ID in lumberjack_loci:
                lumberjack_seq = line.split()[6]
                
                lumber_seqs[tag_ID] = lumberjack_seq
                
    return lumber_seqs
```
