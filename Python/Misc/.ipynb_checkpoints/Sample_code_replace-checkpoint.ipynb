```python
## label changer
import sys

#data_file = open(sys.argv[1], 'r').readlines()
#outfile = open(sys.argv[2], 'w')
#label_keys = open(sys.argv[3], 'r').readlines()

data_file = open("/home/dan/Dropbox/PhD/Dans_PhD_Shared/Thesis/Chapter_4_Hybridisation_and_introgression/data/Microsat_figs_final/All_samples_assignments.txt", 'r').readlines()
label_keys = open("/home/dan/Dropbox/PhD/Dans_PhD_Shared/Thesis/Labels_key_file", 'r').readlines()
outfile = open("/home/dan/Dropbox/PhD/Dans_PhD_Shared/Thesis/Chapter_4_Hybridisation_and_introgression/data/Microsat_figs_final/All_spp_assignments_newcodes.txt", 'w')

for datline in data_file:

    for line in label_keys:
        #print line.split()[0]
        old = line.split()[0]
        new = line.strip().split()[1]
        for field in datline.split():
            if old == field:
                datline = datline.replace(old, new)
            else:
                pass
    outfile.write(datline)
    print datline.strip()
outfile.close()


```
