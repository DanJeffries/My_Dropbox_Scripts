```python
%matplotlib inline
import os
from matplotlib import pyplot as plt
import matplotlib.cm as cmx

```

```python
cm = plt.get_cmap('jet') 
```

```python
directory = "/Users/djeffrie/Desktop/POWSIM_updated/REAL_RUN_BATCH_1/"
```

```python
Fst_dict = {}

Fsts = []
samples = []
Chipowers = []
Fisher_powers = []

for root, dirs, files in os.walk(directory):
    for fil in files:
        if "ou" in fil:
            outfile = open("%s/%s" % (root, fil), 'r').readlines()
            for line in outfile:
                if "Expected Fst" in line:
                    Fst = line.split()[3]
                elif "Diploid sample sizes" in line:
                    sample_number = line.split()[4]
                elif "Tot" in line:
                    Chi = line.split()[4]
                    Fishers = line.split()[6]

        if Fst not in Fst_dict:
            Fst_dict[Fst] = {}
            Fst_dict[Fst]["N_samples"] = []
            Fst_dict[Fst]["Chipowers"] = []
            Fst_dict[Fst]["Fisherpowers"] = []

        Fst_dict[Fst]["N_samples"].append(int(sample_number))
        Fst_dict[Fst]["Chipowers"].append(float(Chi))
        Fst_dict[Fst]["Fisherpowers"].append(float(Fishers))
```

```python
sample_numbers = [9,4,27,13,12,8,21,24,7,37,27,20,14,20,14]
pool_numbers = [82,88,21,20,13,33]
```

```python
help(plt.vlines)
```

```python
Fish_cm = plt.get_cmap('Blues') 
Fish_map = Fish_cm(5)
Chi_cm = plt.get_cmap('Oranges') 
Chi_map = Chi_cm(1)

Chi_col_index = 100
Fish_col_index = 100


fig = plt.figure(figsize = (16,10))

plt.axis([0, 90, 0, 1.1])

Fsts_sorted = sorted(Fst_dict)

for Fst_val in Fsts_sorted:
    
    plt.plot(sorted(Fst_dict[Fst_val]["N_samples"]),sorted(Fst_dict[Fst_val]["Chipowers"]), "-o", label = Fst_val, color = Chi_cm(Chi_col_index))
    plt.plot(sorted(Fst_dict[Fst_val]["N_samples"]),sorted(Fst_dict[Fst_val]["Fisherpowers"]), "-o", label = Fst_val, color = Fish_cm(Fish_col_index))
    
    Fish_col_index += 30
    Chi_col_index += 30

for pop in sample_numbers:
    plt.vlines(pop,0,1.1, color = "lightgrey", linestyles="dashed")
    
for pool in pool_numbers:
    plt.vlines(pool, 0,1.1,color = "black", linewidths = 2, linestyles="dashed")

plt.legend()
plt.xlabel("N samples per population")
plt.ylabel("Power (1-Beta)") # β


fig.savefig("/Users/djeffrie/Desktop/POWSIM_updated/REAL_RUN_BATCH_1/Power_tests.pdf", format = "pdf")
```
