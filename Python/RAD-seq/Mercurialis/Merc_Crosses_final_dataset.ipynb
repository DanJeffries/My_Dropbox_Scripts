## Plots and summary stats for the final RADseq dataset from the M. annua (M) x M. huetii (F) cross. 

### Stacks parameter tests 

Testing the M parameter. I ran Ustacks for all parents and for 14 offspring for M1 - M5. Note that parents are pure, whereas offspring are hybrids, this the M value to use is likely going to by different across the two groups. 

I counted the number of "stacks" output by Stacks (after secondary merging, and removal of repetitive loci).



```python
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure(figsize = (10,8))

with open("/home/djeffrie/Dropbox/Sexy_plants/Merc_crosses_RAD/Stacks/M_tests/M_tests_Nstacks_parents.txt", 'r') as parents:
    for line in parents:
        if not "M1" in line:
            data = [(int(line.split()[i])-int(line.split()[i+1]))/max([int(i) for i in line.split()]) for i in range(0,len(line.split())-1)]
            plt.plot(data, c = "black")
            

with open("/home/djeffrie/Dropbox/Sexy_plants/Merc_crosses_RAD/Stacks/M_tests/M_tests_Nstacks_offspring.txt", 'r') as offspring:
    for line in offspring:
        if not "M1" in line:
            data = [(int(line.split()[i])-int(line.split()[i+1]))/max([int(i) for i in line.split()]) for i in range(0,len(line.split())-1)]
            plt.plot(data, c = "lightgrey")

            
plt.xticks(range(0,len(line.split())-1), ["1>2","2>3","3>4","4>5",])
plt.ylabel("Proportion of max loci found")
plt.title("Effect of M on number of merged loci")
plt.show()
```
