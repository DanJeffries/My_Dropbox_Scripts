```python
fasta_path = "/home/djeffrie/Data/Genomes/M_huetii/wtdbg2/S3_e2_A_p19_cons.fa"
```

```python
from Bio import SeqIO
```

```python
fasta = SeqIO.parse(open(fasta_path, 'r'), "fasta")
```

```python
hist_dat = []
```

```python
for record in fasta:
    hist_dat.append(len(record.seq))
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize = (30,10))
plt.hist(hist_dat, bins= 1000)
plt.xlim(0,1000000)
plt.vlines(246395, 0, 800)
plt.show()
```
