```python
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
blastouts = open("/home/djeffrie/Data/MITObim_Rtemp/MITObim/MITObim_assembly_iteration_85/testing_circularity/Circularity_test.out", 'r').readlines()
```

```python
read_IDs = []
subject_alignement_bounds_list = []

for line in blastouts:
    readID = line.strip().split()[0]
    alignment_idscore = float(line.strip().split()[2])
    alignment_length = float(line.strip().split()[3])
    alignment_start = float(line.strip().split()[8])
    alignment_end = float(line.strip().split()[9])
    subject_alignement_bounds = (alignment_start, alignment_end)
    
    if readID not in read_IDs:
        read_IDs.append(readID)
        if alignment_idscore > 90 and alignment_length > 70:
            subject_alignement_bounds_list.append(subject_alignement_bounds)
    
```

```python
subject_bases = {}
for i in range(1,989,1):
    subject_bases[i] = 0
    
```

```python
for i in subject_alignement_bounds_list:    
    for base in subject_bases:
        if base >= sorted(i)[0] and base <= sorted(i)[1]:
            subject_bases[base] += 1
```

```python
fig = plt.figure(figsize=(20,15))

bar_pos = []
barheight = []

for i in subject_bases:
    bar_pos.append(i)
    barheight.append(subject_bases[i])

plt.bar(bar_pos, barheight, color = "royalblue", edgecolor = 'none')
plt.xticks(range(0,1000,100), range(0,1000,100))
plt.show()

```
