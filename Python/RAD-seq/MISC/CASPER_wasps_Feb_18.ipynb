```python
myvcf = vcf.Reader(open("/home/djeffrie/Data/RADseq/CASPER/sex_asex_for_qtl.vcf", 'r'))

homozygosity_threshold = 1
heterozygosity_threshold = 0

Parth_hom_loci = []

for record in myvcf:
    call_count = 0
    homozygote_count = 0
    heterozygote_count = 0
    Parth_hom = False
    Sex_het = False

    for sample in record.samples:
        if sample.sample in parthenogens:

            if sample.called:
                call_count += 1

                if sample["GT"] == "0/0" or sample["GT"] == "1/1":
                    homozygote_count += 1


    homozygosity = homozygote_count/call_count
    #print "homozygosity = %s " % (homozygosity)

    if homozygosity >= homozygosity_threshold:
        Parth_hom = True
        
    for sample in record.samples:
        if sample.sample in sexuals:

            if sample.called:
                call_count += 1

                if sample["GT"] == "0/1" or sample["GT"] == "1/0":
                    heterozygote_count += 1


    heterozygosity = heterozygote_count/call_count
    #print "homozygosity = %s " % (homozygosity)

    if heterozygosity >= heterozygosity_threshold:
        Sex_het = True
        
    if Parth_hom == True and Sex_het == True:
        print "yaayyyyyy %s" % record.ID
        
    print "Parthenogens homozygosity = %s" % homozygosity
    print "Sexuals heterozygosity = %s" % heterozygosity
```
