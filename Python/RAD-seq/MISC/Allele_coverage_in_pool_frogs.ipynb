## Finding coverage ratios between R and L alleles in hybrid(ogenetic) pool frogs.

Some of the below samples are hypothesised to be triploid. If so their genotypes will either be RRL or RLL and thus, the coverage ratios of the R and L alleles will be either 2:1 or 1:2. 

### Contents:

0. Data summary
1. Find species diagnostic loci
2. Calculate allele ratios at species diagnostic loci
3. Plot coverage ratios



```python
from __future__ import division
import numpy as np
import MISC_RAD_tools as MISC
import vcf
from matplotlib import pyplot as plt
```

## 1. Find species diagnostic loci

I will first find all the loci which are diagnostic between R and L, that is fixed between them. These loci must be heterozygous in hybrids so I will

```python
my_vcf = vcf.Reader(open("/home/djeffrie/Data/Dufresnes_pool_frogs/batch_1.vcf", 'r'))

RRs = ["STA14","STA15","STA16","STA17","STA18"] 
LLs = ["BIA01","BIA02","BIA03","BIA04","BIA05","BIA07","BIA08","BIA09","BIA11","BIA12","BIA14","BIA15","BIA16","BIA18",
       "CAM02","CAM11","CAM15","LOZ01","LOZ03","LOZ04","LOZ05","LOZ06","LOZ08","LOZ09","LOZ10","LOZ11","LOZ14","LOZ15",
       "LOZ17","LOZ19","LOZ21","LOZ22","Godelino4","Godelino5","Joux2","Joux3","Joux4","Joux5","Joux7","Joux8","Gudo1",
       "Gudo3","Parone1","Parone2","Parone3","Parone5","Parone6","Parone7","Novara6","Novara10"]

species_diagnostic = []

count = 1

Species_alleles = {}  ## will contain only species diagnostic loci

for record in my_vcf:
    
    count += 1
    R_alleles = []
    L_alleles = []
    
    for sample in record.samples:
        if sample.sample in RRs:
            if sample.called:
                R_alleles.append(sample.gt_bases[0])
                R_alleles.append(sample.gt_bases[2])
            
        elif sample.sample in LLs:
            if sample.called:
                L_alleles.append(sample.gt_bases[0])
                L_alleles.append(sample.gt_bases[2])
    
    if len(set(R_alleles)) == 1 and len(set(L_alleles)) == 1:

        for allele in R_alleles:
            if allele not in L_alleles:
                if record.ID not in Species_alleles:
                    Species_alleles[record.ID] = {}
                Species_alleles[record.ID]["R"] = allele

        for allele in L_alleles:
            if allele not in R_alleles:
                if record.ID not in Species_alleles:
                    Species_alleles[record.ID] = {}
                Species_alleles[record.ID]["L"] = allele
                
print "Number of species specific loci = %s" % len(Species_alleles)
```

Note - in the summary plot, the average heterozygosity of the hybrids is about 20%. Given that we have about 2500 loci, that means about 500 loci should be heterozygous in each hybrid. So 376 species diagnostic loci out of a potential ~500 is about right I think, given that some of these will be heterozygous in the parents. 

## 2. Calculate allele ratios at species diagnostic loci

Here I look just at the 376 loci found above. For each locus in the RL hybrids, I divide the lowest allele coverage by the highest. For a diploid, the resulting ratio (averaged across all loci for each sample) should be around 1. For triploids which are either RRL or RLL, it should be 2. 

```python
RLs = ["BIA06","BIA10","BIA13","BIA17","BIA19","CAM03","CAM04","CAM05","CAM06","CAM08","CAM09","CAM10","CAM12","CAM13","CAM16","CAM17","GOL01","GOL03","GOL06","GOL07","GOL08",
       "GOL09","GOL10","GOL11","GOL12","LOZ02","LOZ07","LOZ23","Godelino2","Godelino3","Godelino7","MGO01","MGO02","MGO04","MGO06","MGO07","MGO08","MGO09","MGO10","MGO11","MGO12",
       "AGR01","AGR02","AGR03","AGR04","AGR05","AGR06","AGR07","AGR08","AGR09","AGR10","AGR11","AGR14""AGR15","Agra1","Agra3","STA01","STA03","STA04","STA05","STA06","STA07","STA08",
       "STA09","STA10","STA11","STA12","Gudo2","Parone4","Dosolo24","Dosolo26"]



my_vcf = vcf.Reader(open("/home/djeffrie/Data/Dufresnes_pool_frogs/batch_1.vcf", 'r'))

test_sample = "BIA01"


sample_dict = {}

for record in my_vcf:
    
    if record.ID in Species_alleles:

        for sample in record.samples:
            if sample.sample in RLs:
                if sample.sample not in sample_dict:
                    sample_dict[sample.sample] = []

                if 30 > sample["DP"] > 15 and sample.is_het:

                    if not any([sample["AD"][0] > sample["AD"][1]*3, sample["AD"][1] > sample["AD"][0]*3]):
                                               
                        
                        R_depth = None
                        L_depth = None

                        if sample.gt_bases[0] == Species_alleles[record.ID]["R"]:
                            R_depth = sample["AD"][0]
                        elif sample.gt_bases[2] == Species_alleles[record.ID]["R"]:
                            R_depth = sample["AD"][1]

                        if sample.gt_bases[0] == Species_alleles[record.ID]["L"]:
                            L_depth = sample["AD"][0]
                        elif sample.gt_bases[2] == Species_alleles[record.ID]["L"]:
                            L_depth = sample["AD"][1]


                        if not any([R_depth == None, L_depth == None]):

                            R_prop = R_depth/sample["DP"]
                            L_prop = L_depth/sample["DP"]



                            ratio = R_prop/L_prop

                            sample_dict[sample.sample].append(ratio)


            
```

### Output the coverage ratios to a file

```python
Cov_ratios_out = open("/home/djeffrie/Data/Dufresnes_pool_frogs/Allele_coverage_ratios.dat", 'w')

Cov_ratios_out.write("Sample\tMean_cov_ratio\tStDev\n")


for sample in sample_dict:
    Cov_ratios_out.write("%s\t%s\t%s\n" % (sample, np.round(np.mean(sample_dict[sample]),2), np.round(np.std(sample_dict[sample]),2)))

Cov_ratios_out.close()
    
```

## 3. Plot coverage ratios

I will try first just with the RL samples and a handful of known diploids

```python

#define lists below just used for plotting

## Potential triploid hybrids
triploids = ["MGO01","MGO02","MGO04","MGO06","MGO07","MGO08","MGO09","MGO10","MGO11","MGO12","AGR01","AGR02","AGR03","AGR04","AGR05","AGR06","AGR07","AGR08","AGR09","AGR10","AGR11","AGR15","Agra1","Agra3"]

## diploids
diploids = ["BIA06","BIA10","BIA13","BIA17","BIA19"]

fig = plt.figure(figsize = (50,15))         

pos = 1

samples = []
positions = []

for sample in diploids:
    if sample in sample_dict:
        
        ## plot violin
        vio = plt.violinplot(sample_dict[sample], positions = [pos], showmedians=True, showextrema=False)
        
        ## plot the points inside the violin
        plt.scatter([pos]*len(sample_dict[sample]), sample_dict[sample], s = 1, c = "black")
        
        ## print number of loci tested per sample
        plt.text(pos, 0.25, str(len(sample_dict[sample])), horizontalalignment='center', fontsize = 15)
        
        positions.append(pos)
        samples.append(sample)
        
        pos += 1
        
        for pc in vio["bodies"]:
            pc.set_facecolor("grey")
            pc.set_edgecolor("black")
        vio["cmedians"].set_edgecolor("Black")

pos += 1

for sample in triploids:
    if sample in sample_dict:
        vio = plt.violinplot(sample_dict[sample], positions = [pos], showmedians=True, showextrema=False)
        
        ## plot the points inside the violin
        plt.scatter([pos]*len(sample_dict[sample]), sample_dict[sample], s = 1, c = "black")
        
        ## print number of loci tested per sample
        plt.text(pos, 0.25, str(len(sample_dict[sample])), horizontalalignment='center', fontsize = 15)
        
        positions.append(pos)
        samples.append(sample)
        
        pos += 1
        
        for pc in vio["bodies"]:
            pc.set_facecolor("grey")
            pc.set_edgecolor("black")
        vio["cmedians"].set_edgecolor("Black")

plt.xticks(positions, samples, rotation = 45, fontsize = 15)
plt.ylabel("Lowest allele coverage / Highest allele coverage", fontsize = 15)
plt.title("Allele coverage ratio at 376 species diagnostic SNPs", fontsize = 25)
plt.savefig("/home/djeffrie/Data/Dufresnes_pool_frogs/Allele_cov_ratios.pdf")
plt.show()
```
