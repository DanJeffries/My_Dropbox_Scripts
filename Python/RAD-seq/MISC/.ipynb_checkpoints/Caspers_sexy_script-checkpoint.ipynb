### First we try to find loci that are homozygous in parthenogens and heterozygous or homozygous for the other allele in sexuals.


```python
def parth_loc_finder(record, parth_threshold, sexuals_threshold):
    
    N_Homozygous_parths = 0
    n_parths_called = 0
    n_sexuals_called = 0
    Parth_hom_allele = []
    Het_or_alt_hom_sexuals = 0
    N_sexuals_not_hom_for_parth_allele = 0

    ## Is the locus homozygous in all parthenogens? ----------------------------------------------

    for sample in record.samples:
        sample_name = sample.sample
        if sample_name in parthenogens:
            #print sample_name, "called? =", sample.called, "Het? = ", sample.is_het, sample.gt_bases
            if sample.called == True:
                n_parths_called += 1
                #print "%s: %s" % (sample_name, sample.gt_bases)
                if sample.is_het == False:
                    N_Homozygous_parths += 1
                    Parth_hom_allele.append(sample.gt_bases.split("/")[0])
                    
    
    Parth_hom_thresh_limit_exceeded = False ## default state
    
    if N_Homozygous_parths >= n_parths_called*parth_threshold:
        Homozygous_in_parthenogens = True
        #print "%s_%s is homozygous in all parthenogens" % (record.ID, record.POS)

        ## Are all parthenogens homozygous for the same allele?
        
        if len(set(Parth_hom_allele))  == 1: ## if only one allele in parthenogens
            #print "Parthenogens are homozygous for the same allele"
            Parthenogens_hom_for_same_allele = True
            Parthogen_allele = list(set(Parth_hom_allele))[0]
            #
            #print "Parthenogens are homozygous for the same allele (%s)" % Parthogen_allele

        elif len(set(Parth_hom_allele))  > 1:

            major_parth_allele = max(Counter(Parth_hom_allele).values())
            minor_parth_allele = min(Counter(Parth_hom_allele).values())

            if major_parth_allele/n_parths_called >= parth_threshold: 
                Parthenogens_hom_for_same_allele = True
                Parth_hom_thresh_limit_exceeded = True
                Parthogen_allele = max(Counter(Parth_hom_allele).iteritems(), key=operator.itemgetter(1))[0]
                

            else:
                Parthenogens_hom_for_same_allele = False
                
        else:
            Parthenogens_hom_for_same_allele = False
            
    else:
        Homozygous_in_parthenogens = False
        

    ## If the locus is homozygous in parthenogens, is it heterozygous or homozygous for the other allele in sexuals?

    if Homozygous_in_parthenogens == True and Parthenogens_hom_for_same_allele == True:
        #print "Yes looking in sexuals"
        for sample in record.samples:
            sample_name = sample.sample
            if sample_name in sexuals:
                if sample.called == True:
                    n_sexuals_called += 1
                    #print "%s : %s" % (sample_name, sample.gt_bases)
                    
                    if sample.is_het == False: ## if sample is homozygous
                        
                        if not sample.gt_bases.split("/")[0] == Parthogen_allele: ## check it is not homozygous for the same allele as the parthenogens.
                            N_sexuals_not_hom_for_parth_allele += 1
                            

        if N_sexuals_not_hom_for_parth_allele >= n_sexuals_called*sexuals_threshold:
            Het_or_alt_hom_in_sexuals = True
            
            print "\nLocus %s_%s" % (record.ID, record.POS)
            
            if Parthenogens_hom_for_same_allele == True and Parth_hom_thresh_limit_exceeded == False:
                print "Parthenogens are homozygous for the same allele"
            elif Parth_hom_thresh_limit_exceeded == True:
                print "Parthenogens are over the homozygosity threshold for one allele"
            print "Heterozygous or homozygous for alternative allele in Sexuals"
```

```python
### Looking only for alleles that are present in sexuals, and not in parthenogens

def Sexuals_specific_locus_finder(vcf_object, parths_threshold = 0.95, sexuals_threshold = 0.05):
    
    ## parths_threshold - the sexual allele has to be absent from at least this many parthenogens. Default 95%
    ## sexuals_threshold - an allele has to be present in at least this number of sexuals to be counted. Default 5%
    
    sex_spec_allele_loci = []
    
    
    for record in vcf_object:
        
        N_Homozygous_parths = 0
        N_parths_called = 0
        N_parths_without_sex_allele = 0
        N_sexuals_called = 0
        Parth_hom_allele = []
        Het_or_alt_hom_sexuals = 0
        Sex_removed = None
        Parth_removed = None

        ## Get the alleles in the sexual? ----------------------------------------------

        Sexuals_alleles = []
        Parthenogen_alleles = []
        #print "Sexuals"

        for sample in record.samples:
            sample_name = sample.sample
            
            if sample_name in sexuals:
                N_sexuals_called += 1

                if sample.called == True:
                    [Sexuals_alleles.append(i) for i in sample.gt_bases.split("/")] ## both alleles added!
        
          
        
        major_sex_allele = max(Counter(Sexuals_alleles).values())
        minor_sex_allele = min(Counter(Sexuals_alleles).values())
        Major_sex_allele = max(Counter(Sexuals_alleles).iteritems(), key=operator.itemgetter(1))[0]         
        Minor_sex_allele = min(Counter(Sexuals_alleles).iteritems(), key=operator.itemgetter(1))[0]
        
        if minor_sex_allele < N_sexuals_called*sexuals_threshold:
            Sexuals_alleles = [x for x in Sexuals_alleles if x != Minor_sex_allele] # removes minor allele from the list.
            Sex_removed = Minor_sex_allele
        

        for sample in record.samples:
            sample_name = sample.sample
            
            if sample_name in parthenogens:
                
                if sample.called == True:
                    N_parths_called += 1
                    [Parthenogen_alleles.append(i) for i in sample.gt_bases.split("/")]
                    
        major_parth_allele = max(Counter(Parthenogen_alleles).values())
        minor_parth_allele = min(Counter(Parthenogen_alleles).values())
        Major_parth_allele = max(Counter(Parthenogen_alleles).iteritems(), key=operator.itemgetter(1))[0]
        Minor_parth_allele = min(Counter(Parthenogen_alleles).iteritems(), key=operator.itemgetter(1))[0]
        
        if minor_parth_allele < N_parths_called*parths_threshold:
            Parthenogen_alleles = [x for x in Parthenogen_alleles if x != Minor_parth_allele] # removes minor allele from the list.
            Parth_removed = Minor_parth_allele
        shared = True
        
        for allele in set(Sexuals_alleles):
            if not allele in set(Parthenogen_alleles):
                shared = False
        if shared == False:
            print "record %s_%s" % (record.ID, record.POS)
            print "Parthenogen alleles = %s" % set(Parthenogen_alleles)
            if Parth_removed != None:
                print "Allele %s was removed from pathenogens as its freq was below the threshold (%s)" % (Minor_parth_allele, parths_threshold)
                
            print "Sexual alleles = %s" % set(Sexuals_alleles)
            if Sex_removed != None:
                print "Allele %s was removed from sexuals as its freq was below the threshold (%s)" % (Minor_sex_allele, parths_threshold)
                
            print "Sexual specific allele: %s" % (allele)
            sex_spec_allele_loci.append("%s_%s" % (record.ID, record.POS))
                
    return sex_spec_allele_loci

```

```python
vcf_read = vcf.Reader(open(alteredvcfpath, 'r'))

fig = plt.figure(figsize=(15,15))

    
low_aaf_depths = []
high_aaf_depths = []

for record in vcf_read:
    
    record_depths = []
    for sample in record:
        depth = sample['DP']
        record_depths.append(depth)
        
    mean_depth = np.mean(record_depths)

    if record.aaf[0] < 0.05:
        plt.scatter(record.aaf,mean_depth, color = "blue")
        low_aaf_depths.append(mean_depth)

    elif record.aaf[0] >= 0.05:
        plt.scatter(record.aaf,mean_depth, color = "red")
        high_aaf_depths.append(mean_depth)
        

plt.axhline(np.mean(low_aaf_depths), color = "blue")
plt.axhline(np.mean(high_aaf_depths), color = "red")
    
plt.show()
```

So there is a big difference (about half) between the coverage at loci with very low aaf and those with aaf above 0.05. This supports the idea that these loci suffer from too low coverage in some samples and that there is allele dropout in many samples.

```python
def coverage_counter(indir): ## Need to add help info to this and annotate properly!!
    
    """
    
    This function calculates the coverage of each tag in the ustacks tags.tsv output
    as long as that locus has not been blacklisted. It outputs a text file containing all 
    tag IDs and coverage values, as well as a histogram of the tag coverage into the same 
    directory as the input file.
    
    - filename: the full path to the tags.tsv file
    
    This function works in conjuction with coverage_counter_looper and returns a dictionary for this.
    
    """    
    
    import csv
    import os
    import gzip
    import collections
    
    Coverage_dict = {}
    
    for root, dirs, files in os.walk(indir):
        for tags_file in files:
            if "tags.tsv" in tags_file and "catalog" not in tags_file:
                file_name = "%s/%s" % (root, tags_file)
                           

                csvcol3 = []        

                if file_name.endswith(".gz"):
                    csvfile = gzip.open(file_name, 'rb')
                else: 
                    csvfile = open(file_name, 'rb') 

                csvread = csv.reader(csvfile, delimiter ="\t") ## read csv in    
                blacklisted_loci = []

                sample_name = file_name.rpartition("/")[2].split(".")[0]
                
                n_blacklisted = 0
                n_counted = 0

                for line in csvread:
                    if not line[0].startswith('#'): # outputs in new version of stacks start with a comment 
                        tag_ID = line[2]
                        if "consensus" in line and line[11] == "1": ## Check that the locus is not blacklisted!
                            n_blacklisted += 1
                            blacklisted_loci.append(tag_ID)
                        if "model" not in line and "consensus" not in line:
                            n_counted += 1
                            csvcol3.append(tag_ID) ## for each read, add its tag ID to a list.

                coverage_count = collections.Counter(csvcol3) ## count the occurance of each tag ID in the list to get read depths per locus
                good_counts = {}

                ## Now remove coverage counts for loci in the blacklist. 

                for locus in coverage_count: 
                    if locus not in blacklisted_loci:
                        good_counts[locus] = coverage_count[locus]


                ## OUTPUTS -------------------------

                Coverage_dict[sample_name] = good_counts
                
    mean_covs = {}
    for sample in Coverage_dict:
        mean_covs[sample] = np.mean(Coverage_dict[sample].values())
    print "Done"
    return Coverage_dict, mean_covs
```

```python
def Tag_counter(indir):
    
    import csv
    import os
    import gzip
    import collections
    
    Tag_counts = {}
    
    for root, dirs, files in os.walk(indir):
        for tags_file in files:
            if "tags.tsv" in tags_file and "catalog" not in tags_file:
                file_name = "%s/%s" % (root, tags_file)
                sample_name = file_name.rpartition("/")[2].split(".")[0]

                if file_name.endswith(".gz"):
                    csvfile = gzip.open(file_name, 'rb')
                else: 
                    csvfile = open(file_name, 'rb') 

                csvread = csv.reader(csvfile, delimiter ="\t") ## read csv in    
                
                n_tags = 0

                for line in csvread:
                    if not line[0].startswith('#'): # outputs in new version of stacks start with a comment 
                        tag_ID = line[2]
                        if "consensus" in line and not line[11] == "1": ## Check that the locus is not blacklisted!
                            n_tags += 1
                Tag_counts[sample_name] = n_tags
    print "Done"
    
    return Tag_counts
```

```python
indir = "/home/djeffrie/Data/Caspers_data/tags_files/"


def coverage_counter(indir, verb): ## Need to add help info to this and annotate properly!!
    
    """
    
    This function calculates the coverage of each tag in the ustacks tags.tsv output
    as long as that locus has not been blacklisted. It outputs a text file containing all 
    tag IDs and coverage values, as well as a histogram of the tag coverage into the same 
    directory as the input file.
    
    - filename: the full path to the tags.tsv file
    
    This function works in conjuction with coverage_counter_looper and returns a dictionary for this.
    
    """    

    csvcol3 = []        
    
    if file_name.endswith(".gz"):
        csvfile = gzip.open(file_name, 'rb')
    else: 
        csvfile = open(file_name, 'rb') 
    
    csvread = csv.reader(csvfile, delimiter ="\t") ## read csv in    
    blacklisted_loci = []
    
    sample_name = file_name.rpartition("/")[2].split(".")[0]
    tags_file_parameter = file_name.rpartition("/")[0].rpartition("/")[2].split("_")[0]
    tags_file_val = file_name.rpartition("/")[0].rpartition("/")[2].split("_")[1]
    
    
    n_blacklisted = 0
    n_counted = 0
    
    for line in csvread:
        if not line[0].startswith('#'): # outputs in new version of stacks start with a comment 
            tag_ID = line[2]
            if "consensus" in line and line[11] == "1": ## Check that the locus is not blacklisted!
                n_blacklisted += 1
                blacklisted_loci.append(tag_ID)
            if "model" not in line and "consensus" not in line:
                n_counted += 1
                csvcol3.append(tag_ID) ## for each read, add its tag ID to a list.
    
    coverage_count = collections.Counter(csvcol3) ## count the occurance of each tag ID in the list to get read depths per locus
    good_counts = {}
    
    ## Now remove coverage counts for loci in the blacklist. 
    
    for locus in coverage_count: 
        if locus not in blacklisted_loci:
            good_counts[locus] = coverage_count[locus]
    
    
    ## OUTPUTS -------------------------
    
    outdir = file_name.rpartition("/")[0] ## same directory as the input file
    f = open("%s/Coverage data.txt" % outdir, 'w') ## txt file for coverage data    
        
    for i,j in good_counts.iteritems():
        f.write("%s\t%s\n" % (str(i), str(j) )) ## write the coverage data to the txt file
    
    ## Can make individual histograms for each sample if you want. Just un comment this section ------
    
    x = good_counts.values()
    #fig = plt.figure()
    #ax = fig.add_subplot(111)
    #ax.hist(x, bins=max(x), range = (1, np.mean(x)+(2*np.std(x))))
    #ax.set_title("%s, %s = %s" % (sample_name, tags_file_parameter, tags_file_val), fontsize = 15) ####
    #ax.set_xlabel("Coverage", fontsize = 12)
    #ax.set_ylabel("Frequency", fontsize = 12)
    #ax.text(0.6, 0.95, "Mean tag coverage = %s (+/- %s)" % (np.round(np.mean(x),2),np.round(np.std(x),2)) , transform=ax.transAxes, fontsize = 13)
    #plt.savefig("%s/Coverage.pdf" % outdir)
    
    #plt.show()
    #plt.close()
    
    f.close()
    
    if verb == True:
        print "Mean and SD of coverage for %s, %s=%s:  %s (+/- %s)\n" % (sample_name, tags_file_parameter ,tags_file_val , np.round(np.mean(x),2),np.round(np.std(x),2))
    
    return good_counts

```

### Another thing to try is mapping to the genome. 
The idea is to 
1. Map only SNPs that are heterozygous in sexuals
2. Map only SNPs heterozygous in parthenogens
3. Look for a large region of the genome which has a lot of differences between the two mappings. As parthenogens should be homozygous for parthenogenetically linked loci, this regions, this region could be the parthenogenisis locus.

The plan here is to find loci which are heterozygous in sexuals and loci that are heterozygous in parthenogens, and then map them to the genome. The thinking is that the parthenogenetic locus must be homozygous in parthenogens, so there should be a bunch of loci where sexuals have SNPs but parthenogens don't, and they should all be close to eachother. If I map SNPs from both groups to the genome, there should be a region of missing loci in the parthenogens. 

```python
def locus_grouper(vcf_object, N_hetero_sexuals_thresh,N_hetero_parthenogens_thresh):

    Loci_het_in_sexuals = []
    Loci_het_in_parthenogens = []
    Loci_fixed_between_sexuals_and_parthenogens = []
    heterozygous_sexuals = []
    heterozygous_parthenogens = []
    
    
    ## Find loci heterozygous in each group
    for record in vcf_object:
        n_sexuals_called = 0
        n_parthenogens_called = 0
        hetero_sexuals = 0
        hetero_parthenogens = 0
        sexuals_allele = []
        parthenogens_allele = []    
        
        for sample in record.samples:
            if sample.sample in sexuals:
                if sample.called == True:
                    n_sexuals_called += 1
                    
                    if sample.is_het:
                        
                        hetero_sexuals += 1
                        heterozygous_sexuals.append(sample.sample)
                    elif sample.is_het == False:
                        sexuals_allele.append(sample.gt_bases.split("/")[0])

            elif sample.sample in parthenogens:
                if sample.called == True:
                    n_parthenogens_called += 1
                    
                    if sample.is_het:
                        
                        hetero_parthenogens += 1
                        heterozygous_parthenogens.append(sample.sample)
                    elif sample.is_het == False:
                        parthenogens_allele.append(sample.gt_bases.split("/")[0])

        if len(set(sexuals_allele)) == 1 and len(set(parthenogens_allele)) == 1: ## looks for strictly fixed loci only
            if not sexuals_allele[0] == parthenogens_allele[0]:
                Loci_fixed_between_sexuals_and_parthenogens.append("%s_%s" % (record.ID, record.POS))
                #print "%s_%s fixed between groups" % (record.ID, record.POS)

        if hetero_sexuals >= n_sexuals_called*N_hetero_sexuals_thresh:
            Loci_het_in_sexuals.append("%s_%s" % (record.ID, record.POS))
            

        if hetero_parthenogens >= n_parthenogens_called*N_hetero_parthenogens_thresh:
            Loci_het_in_parthenogens.append("%s_%s" % (record.ID, record.POS))    


    return Loci_het_in_sexuals, Loci_het_in_parthenogens, Loci_fixed_between_sexuals_and_parthenogens, heterozygous_sexuals, heterozygous_parthenogens
```

So I have set a heterozygosity threshold of 20% for parthenogens, that is, for a locus to be classed as heterozygous in parthenogens a SNP must be found in at least 20% of them. And I have set a threshold of 10% for the sexuals. This threshold in the parthenogens is very important, 20% means that one or two individuals can be heterozygous for a locus but it still not be counted as such. This allows for a bit of error in the classification of samples as parthenogens or not. 


```python
vcf_read = vcf.Reader(open(alteredvcfpath, 'r'))
sex_het_loci, parth_het_loci, fixed_loci, het_sexuals, het_parthenogens = locus_grouper(vcf_read, 0.1, 0.2)

print "Number of loci heterozygous in sexuals = %s" % len(sex_het_loci)
print "Number of loci heterozygous in parthenogens = %s" % len(parth_het_loci)
print "Number of loci fixed between sexuals and parthenogens = %s" % len(fixed_loci)
#for sample in Counter(het_sexuals):
#    print sample, Counter(het_sexuals)[sample]
#for sample in Counter(het_parthenogens):
#    print sample, Counter(het_parthenogens)[sample]
```
