```python
from collections import Counter

def Diagnostic_snp_finder(working_dir, population_pair): ## working dir = string, population_pair = tuple
    pop_1 = population_pair[0]
    pop_2 = population_pair[1]
    fsts = open(working_dir+"/batch_1.fst_"+pop_1+"-"+pop_2+".tsv", 'r').readlines()
    vcf = open(working_dir+"Ho_filtered_altered.vcf", 'r').readlines()
    
    high_fst_snps = []
    tag_IDs_only = []
    for line in fsts:
        if not line.startswith("#"):
            Loc_ID = line.split()[1]
            scaffold = line.split()[4]
            base = line.split()[5]
            fst = float(line.split()[8])
            
            if fst > 0.95: ## Set FST cut-off
                high_fst_snps.append((Loc_ID, scaffold, base))
                tag_IDs_only.append(Loc_ID)
                print (Loc_ID, scaffold, base), fst
    
    ## Now filter VCF for these and output to filtered vcf
    
    all_high_fst_vcf = open(working_dir+"/All_diagnostic_snps.vcf", 'w')
    tag_IDs_only = []
    for line in vcf:
        if line.startswith("#"):
            all_high_fst_vcf.write(line)
        elif not line.startswith("#"):
            Loc_ID = line.split()[2]
            scaffold = line.split()[0]
            base = line.split()[1]
            
            if (Loc_ID, scaffold, base) in high_fst_snps:
                all_high_fst_vcf.write(line)
                tag_IDs_only.append(Loc_ID)
        
    
    ## Find tags that have more than one snp that is diagnostic on them and write to file
    
    tag_counts = Counter(tag_IDs_only)
    
    super_diagnostic_tags_2 = []
    super_diagnostic_tags_3 = []
    super_diagnostic_tags_4 = []
    super_diagnostic_tags_5 = []
    super_diagnostic_tags_6 = []
    super_diagnostic_tags_7 = []
    
    super_diagnostic_vcf_2 = open(working_dir+"Super_diagnostic_tags_2.vcf", 'w')
    super_diagnostic_vcf_3 = open(working_dir+"Super_diagnostic_tags_3.vcf", 'w')
    super_diagnostic_vcf_4 = open(working_dir+"Super_diagnostic_tags_4.vcf", 'w')
    super_diagnostic_vcf_5 = open(working_dir+"Super_diagnostic_tags_5.vcf", 'w')
    super_diagnostic_vcf_6 = open(working_dir+"Super_diagnostic_tags_6.vcf", 'w')
    super_diagnostic_vcf_7 = open(working_dir+"Super_diagnostic_tags_7.vcf", 'w')
    

    for locus in (tag_counts).iteritems():
        #print locus
        if locus[1] > 1:
            super_diagnostic_tags_2.append(locus[0])            
        if locus[1] > 2:
            super_diagnostic_tags_3.append(locus[0])
        if locus[1] > 3:
            super_diagnostic_tags_4.append(locus[0])
        if locus[1] > 4:
            super_diagnostic_tags_5.append(locus[0])
        if locus[1] > 5:
            super_diagnostic_tags_6.append(locus[0])
        if locus[1] > 6:
            super_diagnostic_tags_7.append(locus[0])

            
    for line in vcf:
        if line.startswith("#"): ## headers
            super_diagnostic_vcf_2.write(line)
            super_diagnostic_vcf_3.write(line)
            super_diagnostic_vcf_4.write(line)
            super_diagnostic_vcf_5.write(line)
            super_diagnostic_vcf_6.write(line)
            super_diagnostic_vcf_7.write(line)
        else:
            Loc_ID = line.split()[2]
            scaffold = line.split()[0]
            base = line.split()[1]
            if (Loc_ID, scaffold, base) in high_fst_snps:
                if Loc_ID in super_diagnostic_tags_2:
                    super_diagnostic_vcf_2.write(line)
                if Loc_ID in super_diagnostic_tags_3:
                    super_diagnostic_vcf_3.write(line)
                if Loc_ID in super_diagnostic_tags_4:
                    super_diagnostic_vcf_4.write(line)
                if Loc_ID in super_diagnostic_tags_5:
                    super_diagnostic_vcf_5.write(line)
                if Loc_ID in super_diagnostic_tags_6:
                    super_diagnostic_vcf_6.write(line)
                if Loc_ID in super_diagnostic_tags_7:
                    super_diagnostic_vcf_7.write(line)
            
    super_diagnostic_vcf_2.close()
    super_diagnostic_vcf_3.close()
    super_diagnostic_vcf_4.close()
    super_diagnostic_vcf_5.close()
    super_diagnostic_vcf_6.close()
    super_diagnostic_vcf_7.close()
    
    print "tags containing 1 or more snps", len(tag_counts)
    print "tags containing 2 or more snps", len(super_diagnostic_tags_2)
    print "tags containing 3 or more snps", len(super_diagnostic_tags_3)
    print "tags containing 4 or more snps", len(super_diagnostic_tags_4)
    print "tags containing 5 or more snps", len(super_diagnostic_tags_5)
    print "tags containing 6 or more snps", len(super_diagnostic_tags_6)
    print "tags containing 7 or more snps", len(super_diagnostic_tags_7)
    
```
