```python
import os
```

```python
def pstacks_script_maker(bam_directory1):
    f = open('./pstacks_scripts.sh', 'w')
    f.write('#!bin/bash\n')
    for root, dirs, files in os.walk(bam_directory1):
        ID = 1
        for fil in files:
            if 'uniq' in fil and 'bai' not in fil and 'bam' not in fil:
                print 'gzip -d '+bam_directory1+fil+' ; pstacks -t sam -f '+bam_directory1+fil[:-3]+' -o ./pstacks_all/ -i '+str(ID)+' -m 10 -p 7 \n'
                f.write('gzip -d '+bam_directory1+fil+' ; pstacks -t sam -f '+bam_directory1+fil[:-3]+' -o ./pstacks_all/ -i '+str(ID)+' -m 10 -p 7 \n')
                ID += 1
        print ID
                
                
            
pstacks_script_maker('./BWA_mapping_batch1/')
```
