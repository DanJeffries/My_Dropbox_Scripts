```python
import os
import re
```

## Make pop_codes file

```python
samples_names = []
new_pop_codes = []
newest_pop_codes = []

f = open('/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Stacks/Stacks_analyses_TRIMMED/Pure_cru_only/pop_codes.txt', 'r')
for line in f.readlines():
    samples_names.append(line.split()[0].rpartition('.')[0])
samples_set = sorted(set(samples_names))

for sample in samples_set:
    new_pop_codes.append(sample+'\t'+re.sub(u'\d','', sample.partition('_')[0]))


nf = open('/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Stacks/Stacks_analyses_TRIMMED/POP_codes.txt', 'w')
for line in new_pop_codes:
    nf.write(line+'\n')
nf.close() 
```

## Make scripts

```python
def Cstacks_script_maker(parent_dir, output_dir, pop_file):
    samples = []
    pop_list = []
    samples_ready = []
    cwd = parent_dir
    
    pops = open(pop_file, 'r').readlines()
    for pop in pops:
        pop_list.append(pop.strip().split()[1])
    print 'Populations for catalog = ', set(pop_list) , 'N populations = '+str(len(set(pop_list)))
    
    for root, dirs, files in os.walk(parent_dir):
        for fil in files:
            print fil
            for population in pop_list: 
                if 'tags.tsv' in fil:
                    if population in fil:
                       samples.append(fil.rpartition('.')[0].rpartition('.')[0])
    samples = sorted(set(samples))
    
    for sample in samples:
        samples_ready.append(' -s '+parent_dir+sample)
    samples = ''.join(samples_ready)
    
    Ccommand = 'cstacks -b 1 -n 5 -p 7'+samples+' -o '+output_dir
    
    f = open(output_dir+'Cstacks_NEU_cru_only.sh','w')
    f.write('#!/bin/bash\n##working directory = '+str(cwd)+'\n')
    for command in Ccommand:
        f.write(command)
    f.close()
    
    return Ccommand


    
Cstacks_script_maker('/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Stacks/Stacks_analyses_TRIMMED/Pure_cru_only/',"/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Stacks/Stacks_analyses_TRIMMED/Pure_cru_only/", '/media/dan/34D5D1CE642D7E36/2013076_Hanfling_Bernd/Stacks/Stacks_analyses_TRIMMED/Pure_cru_only/pop_codes.txt')
```
