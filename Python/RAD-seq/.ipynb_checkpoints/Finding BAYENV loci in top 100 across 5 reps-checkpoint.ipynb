```python
import os
from collections import defaultdict

## First get the paths of the top100 files - these I got from my BAYENV_results_filter.sh script

file_paths = []
for root, dirs, files in os.walk("/home/dan/RAD_programs/BAYENV/bayenv_release/bottlenecked"):
    if "500000" in root:
        for fil in files:
            if fil == "top100.txt":
                file_paths.append(root+"/"+fil)
                #print (root+"/"+fil)

## Put values into dictionary {loc_ID: list of BFs}

winners_dict = dict()
print winners_dict

for path in file_paths:
    print path
    winners = open(path, 'r').readlines()
    for loc in winners:
        #print loc
        ID = loc.split()[0]
        #print ID
        BF = loc.split()[1]
        #print BF
        if ID in winners_dict.keys():
            #print ID, "already in dict"
            winners_dict[ID].append(float(BF))
        else:
            winners_dict[ID]=[float(BF)]

## get the loci that are present in 3 or more top 100's, average their BFs and output to file    

Winning_loc_BFs = open("/home/dan/RAD_programs/BAYENV/bayenv_release/bottlenecked/Winning_loc_BFs.txt", 'w')
Winning_loc_BFs.write("Loc_ID\tAveraged_BF\tNumber_of_reps_in_top_100\n")

WINNER_count = 0
WINNER_loc_names =[] 
for key in winners_dict.keys():
    if len(winners_dict[key]) >= 3:
        Winning_loc_BFs.write(key+"\t"+str(sum(winners_dict[key])/(len(winners_dict[key])))+"\t"+str(len(winners_dict[key]))+"\n")
        print key, winners_dict[key] #print winners_dict
        WINNER_count +=1
print WINNER_count

Winning_loc_BFs.close()
    
```
