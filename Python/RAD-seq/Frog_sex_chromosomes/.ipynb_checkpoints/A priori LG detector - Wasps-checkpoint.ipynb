```python
%matplotlib inline
from matplotlib import pyplot as plt
import pylev2 as lev ## calculates the Levenshtein distance between two strings
from tqdm import tnrange, tqdm_notebook ## for the progress bar
import vcf
```

```python
input_file = "/home/djeffrie/Data/Linkage_group_detector/Caspers_wasp_vcfs/batch_2h.vcf"

## Alter vcf 

myvcf = open(input_file, 'r').readlines()
alteredvcfpath = "%s%s" %(input_file, ".altered")
alteredvcf = open(alteredvcfpath, 'w')

for line in myvcf:
    if "Allele Depth" not in line:
        alteredvcf.write(line)
    elif "Allele Depth" in line:            
        line = '##FORMAT=<ID=AD,Number=.,Type=Integer,Description="Allele Depth">\n'

        alteredvcf.write(line)
alteredvcf.close()

test_vcf = vcf.Reader(open(alteredvcfpath, 'r'))

## Find number of loci

print "Getting vcf info\n"

record_counter = 0
for record in tqdm_notebook(test_vcf, desc = "Counting loci"):
    record_counter += 1
print "Done"

test_vcf = vcf.Reader(open(alteredvcfpath, 'r'))

print "Now getting sample heterozygosity info"
locus_dict = {}
bad_locus_counter = 0

for record in tqdm_notebook(test_vcf, total=record_counter, desc = "Getting GT patterns"):
    if record.call_rate >= 0.75 and record.aaf[0] > 0.2:
        
        rec_name = "%s_%s" % (record.ID,record.POS)
        
        locus_dict[rec_name] = ""

        for sample in record.samples:
            
            if not sample.called:
                locus_dict[rec_name] += "X"
            else:
                locus_dict[rec_name] += str(sample.gt_type)
            #elif sample.is_het:
            #    locus_dict[rec_name] += "1"
            #elif not sample.is_het:
            #    locus_dict[rec_name] += "0"
    else:
        bad_locus_counter += 1
        
print bad_locus_counter 

```

Because the sexuals can be homozygous or heterozygous, and because loci in sexuals could recombine, it is probable that the parth locus would differ in its genotypes across sexuals. So, one way to get around this is to change all "2s" in this dataset (i.e. the code for homozygous for minor allele), to "1s" (code for het). That way the clustering algorithm will think that these are the same category.

```python
for locus in locus_dict:
    locus_dict[locus].replace("2", "1")
```

```python
locus_Ids = locus_dict.keys()

mismatches = 4

clusters = {}

## Make first gen_seq the first centroid

clusters[locus_Ids[0]] = {}
clusters[locus_Ids[0]]["centroid"] = locus_dict[locus_Ids[0]]
clusters[locus_Ids[0]]["matches"] = []

locus_Ids.remove(locus_Ids[0])

for locus in tqdm_notebook(locus_Ids, total = len(locus_Ids), desc = "Clustering loci"):
    added = False
    for cluster in clusters:
        if lev.levenshtein(locus_dict[locus], clusters[cluster]["centroid"]) <= mismatches:
            clusters[cluster]["matches"].append(locus)
            added = True
    if added == False:
        clusters[locus] = {}
        clusters[locus]["centroid"] = locus_dict[locus]
        clusters[locus]["matches"] = []
            
        
```

### All data (batch_2f)

```python
good_clusters = {}
for i  in clusters:
    if len(clusters[i]["matches"]) > 10:
        good_clusters[i] = len(clusters[i]["matches"])
        #print i, len(clusters[i]["matches"])
        #print clusters[i]["matches"]
        
plt.figure(figsize = (18,10))
plt.bar(range(len(good_clusters)), good_clusters.values(), width=0.6)
plt.xticks(range(len(good_clusters)), good_clusters.keys(), rotation = 90)
plt.show()
```
