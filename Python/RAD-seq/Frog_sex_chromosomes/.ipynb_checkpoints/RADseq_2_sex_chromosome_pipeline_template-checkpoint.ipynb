```python
## INPUT INCREMENTAL PARAMETERS HERE -----------------------------------------

parameters = ["M", "m", "MS"]                                             ## List, eg ["M", "m", "MS"]. If only testing 1 param, this must still be a list e.g. ["M"]
M_vals =   [2,3,4,5,6]                                              ## List, eg range(1,5,1) this would give [1,2,3,4]
m_vals =   [2,3,4,5,6]                                              ## List, eg range(1,5,1) this would give [1,2,3,4]
MS_vals =  [2,3,4]                                                 ## List, eg range(1,5,1) this would give [1,2,3,4]
threads =  7                                                        ## Int. eg 7
parent_dir = "%s/%s" % (working_directory, "Incremental_tests/" )    ## Absolute Dir containing the demultiplexed raw read fasta/fastq files
run_cline_switch = 0                                             ## to switch on (1) to run stacks cline or off (0) to run pipeline on existing outputs

print parent_dir

## Making input dictionary --------------------------------------

Setup_dict = {}
Setup_dict["parameters"] = {}
Setup_dict["threads"] = threads
Setup_dict["parent_directory"] = parent_dir
Setup_dict["run_cline_switch"] = run_cline_switch
Setup_dict["verbose"] = True

for param in parameters:
    if "M" in param:
        Setup_dict["parameters"][param] = M_vals
    if "m" in param:
        Setup_dict["parameters"][param] = m_vals
    if "MS" in param:
        Setup_dict["parameters"][param] = MS_vals

#pp = pprint.PrettyPrinter(indent=2)
#pp.pprint(Setup_dict)
```

```python
Incremental_U_outputs = Inc.IncreMental_U(Setup_dict)
```

### Step 5. Run final Ustacks commands -------------------------------------------------------------------------------------------------------------------

```python
for fil in os.listdir("%s/Incremental_tests/" % working_directory):
    if "fq" in fil:
        print fil
```

```python
### Put final Ustacks parameters here.

Final_M = 3
Final_m = 3
Final_MS = 3
p = 7
Ustacks_out_directory = "%s/Cstacks_tests/" % working_directory


### ------------------------------------------------------

if not os.path.exists(Ustacks_out_directory):
    try:
        os.makedirs(Ustacks_out_directory)
    except:
        os.mkdir(Ustacks_out_directory)

samples = []
ID = 1
sample_IDs = []

for fil in os.listdir("%s/Incremental_tests/" % working_directory):
    if "fq" in fil or "fastq" in fil:
        if fil.endswith("fq"):
            file_type = "fastq"
        elif fil.endswith("fq.gz"):
            file_type = "gzfastq"
        Final_Ustacks_cline = "ustacks -f %s -t %s -i %s -M %s -m %s --max_locus_stacks %s -r -d -p %s -o %s" % ("%s/Incremental_tests/%s" % (working_directory,fil), file_type, ID, Final_M, Final_m, Final_MS, p, Ustacks_out_directory) 

        ## Run the command line and capture logs in log file
        #log = open("%s/Final_Ustacks.log" % Ustacks_out_directory, 'w')
        #subprocess.call(Final_Ustacks_cline.split(), stdout=log, stderr=log )

        sample_IDs.append((fil.rpartition("/")[2],ID))
        ID += 1
    
print "Final ustacks run done, outputs in %s" % Ustacks_out_directory
print "\nSample IDs used:"

for i in sample_IDs:
    print "%s\t%s" % (i[0], i[1])
```
