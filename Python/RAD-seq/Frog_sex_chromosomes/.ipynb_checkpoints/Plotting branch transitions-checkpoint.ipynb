```python
## get the ages of the nodes at the beginning of the branches

node_heights = open("/home/djeffrie/Dropbox/My_Dropbox_Scripts/R/Anc_states_Rana/Node_heights.txt", 'r').readlines()## i.e. the time at the start and end of the branch

node_height_dict = {}
for branch in node_heights[1:]:
    node_height_dict[branch.split()[0]] = (float(branch.split()[1]), float(branch.split()[2]))

node_height_dict
```

```python
transitions = open("/home/djeffrie/Dropbox/My_Dropbox_Scripts/R/Anc_states_Rana/transitions1_clean.txt", 'r').readlines()

## Make per branch dict of transitions

transitions_dict = {}

for i in transitions:
    
    if i.split()[0] == "edge":
        edge = i.split()[1]
        
    if i.split()[0] == "tree":
        tree = i.split()[1]
        if tree not in transitions_dict:
            transitions_dict[tree] = {}
    
    if i.split()[0].startswith("C"):
        Chr = i.split()[0]
        time = i.split()[1]
    
    if i.split()[0] == "x":
        edge = Chr = time = None
    
    if all([edge, Chr, time]):
        if edge not in transitions_dict[tree]:
            transitions_dict[tree][edge] = []
        transitions_dict[tree][edge].append((Chr, float(time)))
    

#transitions_dict
        
```

```python
import numpy as np

cumul_transition_dict = {}

for tree in transitions_dict:
    
    if tree not in cumul_transition_dict:
        cumul_transition_dict[tree] = {}
    
    for edge in transitions_dict[tree]:
        
        segment_count = 0  ## track number of segments in branch
        start_time = node_height_dict[edge][0]  ## get the start time of the branch
        cumul_time = []  ## track time along the branch
        
        
        for segment in transitions_dict[tree][edge]:
            segment_count += 1 
            chrom = segment[0][1]
            time = segment[1]
            
            if segment_count == 1:
                cumul_time += [0]*int(np.round(start_time*10))
                
            cumul_time += [int(chrom)]*int(np.round(time*10))
            
        cumul_transition_dict[tree][edge] = cumul_time
            
```

### make a per branch dictionary, with all transitions in it, converted into slices of time along the branch

```python
per_branch_dict = {}

for i in range(1,53):
    per_branch_dict[i] = []
    
    for tree in cumul_transition_dict:
        
        per_branch_dict[i].append(cumul_transition_dict[tree][str(i)])
        
#print per_branch_dict

    
```

```python
hist_dict = {}

Chroms = [1,2,3,5,8]

for branch in per_branch_dict:
    
    #print branch

    if branch not in hist_dict:
        hist_dict[branch] = {}
        hist_dict[branch][1] = [0]*(len(per_branch_dict[branch][0])+5)
        hist_dict[branch][2] = [0]*(len(per_branch_dict[branch][0])+5)        
        hist_dict[branch][3] = [0]*(len(per_branch_dict[branch][0])+5)        
        hist_dict[branch][5] = [0]*(len(per_branch_dict[branch][0])+5)
        hist_dict[branch][8] = [0]*(len(per_branch_dict[branch][0])+5)
        
        #print hist_dict[branch][1]
        
    for branch_list in per_branch_dict[branch]:
        #print branch, len(branch_list)
        
        for chrom in Chroms:
            #print branch, chrom
            for i in range(0, len(branch_list)):
                if branch_list[i] == chrom:
                    #print i
                    hist_dict[branch][chrom][i] += 1
                
```

Ok now I want to plot the pie charts for each node, based on the proportions of simulations at the start end of each internal branch.

### So I have a dictionary where each key is a branch, and within each branch key there are 5 lists, one for each of the chromosomes, with the counts for the numbers of times that state is found on that branch, across all of the trees in the stochastic mapping

```python
branch_pie_dict = {}

for branch in hist_dict:
    
    
    
    if branch not in branch_pie_dict:
        branch_pie_dict[branch] = {}
    
    for state in hist_dict[branch]:
        
        time_found = False
        N_slices = 0
        N_non_zero_slices = 0
        
        for time_slice in hist_dict[branch][state]:
            
            if time_slice != 0:
                N_non_zero_slices += 1
            
            if N_non_zero_slices == 1:
                branch_pie_dict[branch][state] = hist_dict[branch][state][N_slices]
                time_found = True
                
            N_slices += 1
        
        if not time_found:
            branch_pie_dict[branch][state] = 0
            

        print branch, state, branch_pie_dict[branch][state]
```
