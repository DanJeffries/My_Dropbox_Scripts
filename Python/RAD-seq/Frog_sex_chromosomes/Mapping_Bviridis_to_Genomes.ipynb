### Mapping B. viridis sex linked and linkage group markers to the Xenopus genome, via the Nanorana genome.

First, need to get markers from the LGs in the map from the stacks catalog

To do this I need the map file, the vcf and the catalog file. 


```python
import vcf
import gzip
```

```python
map_file = open("/home/djeffrie/Data/RADseq/STOECK/Bviridis/Linkage_maps/Combined/LOD_3_4/MAP_LOD_3_ordered1.txt", 'r').readlines()
vcf_file = vcf.Reader(open("/home/djeffrie/Data/RADseq/STOECK/Bviridis/Linkage_maps/Combined/batch_1.vcf.altered", 'r'))
#catalog_file = gzip.open("/home/djeffrie/Data/RADseq/STOECK/Bviridis/Linkage_maps/Combined/batch_1.catalog.tags.tsv.gz", 'r').readlines()
```

```python
LG_marker_numbers = []

Female_LGs = {}
Male_LGs = {}

for line in map_file:
    
    if line.startswith("#***") or line.startswith("***") :
        
        LG = line.split()[3]
        Male_LGs[LG] = []
        Female_LGs[LG] = []
    
    elif not line.startswith("#"):
        
        marker_number = line.split()[0]
        male_pos = float(line.split()[1])
        female_pos = float(line.split()[2])

        Male_LGs[LG].append((marker_number, male_pos))
        Female_LGs[LG].append((marker_number, female_pos))
        
LG_of_interest = "1"

for LG in Male_LGs:
    if LG == LG_of_interest:
        
        LG_marker_numbers = [int(i[0]) for i in Male_LGs[LG]]

#LG_marker_numbers
```

```python
LG_marker_IDs = []

record_N = 1

for record in vcf_file:

    if record_N in LG_marker_numbers:
        LG_marker_IDs.append(record.ID)
    
    record_N += 1
    
        
```

```python
## Now make fasta from the catalog file

catalog_file1 = open("/home/djeffrie/Data/RADseq/STOECK/Bviridis/Genome_mapping/Nanorana_genome/Sex_linked_markers/batch_1.catalog.tags.tsv").readlines()

LG_fasta = open("/home/djeffrie/Data/RADseq/STOECK/Bviridis/Genome_mapping/Nanorana_genome/Sex_linked_markers/LG%s_markers.fasta" % LG_of_interest, 'w')

for tag in catalog_file1:
    
    if not tag.startswith("#"):
        tag_ID = tag.split()[2]
        seq = tag.split()[8]
        if tag_ID in LG_marker_IDs:
            LG_fasta.write(">%s_LG_%s\n%s\n" % (tag_ID, LG_of_interest, seq))
LG_fasta.close()
```
