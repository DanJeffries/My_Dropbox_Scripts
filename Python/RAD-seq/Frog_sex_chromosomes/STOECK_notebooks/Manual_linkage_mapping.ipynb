```python
from __future__ import division
from collections import Counter
```

```python
vcf = open("/Users/danieljeffries/Data/RADseq/STOECK/batch_1.vcf" , 'r').readlines()
```

```python
gt_patterns = []
for i in vcf:
    if not i.startswith('#'):
        j = i.split()[9:]
        gts = [h.split(':')[0] for h in j]
        #print gts
        
        #all([any([gts[15] == "1/0", gts[15] == '0/1']), any([gts[16] == "0/0", gts[16] == "1/1"])]) 
        
        if all([any([gts[15] == "1/1", gts[15] == '0/0']), any([gts[16] == "1/0", gts[16] == "0/1"])]):
            #print gts
        
            newgts = []
            missing = 0
            N = 0
            for gt in gts:
                N += 1
                if gt == "0/0":
                    newgts.append("1")
                elif gt == "1/1":
                    newgts.append("2")
                elif gt == "1/0":
                    newgts.append("3")
                elif gt == "0/1":
                    newgts.append("3")
                elif gt == "./.":
                    newgts.append("0")
                    missing += 1
            
            if missing/N < 0.25:
                #print missing/N
                #print newgts
                gt_patterns.append("".join(newgts))

        
for i in sorted(gt_patterns):
    print i
            
                
```
