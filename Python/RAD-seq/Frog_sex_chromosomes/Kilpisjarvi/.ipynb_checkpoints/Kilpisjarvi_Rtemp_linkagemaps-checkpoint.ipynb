# <b>Kilpisjarvi temporaria.</b>

Based on Melâ€™s transcroptome data, there are male vs female Fst patterns on Chr01 as expected and on Chr08 as well in this northern Swedish population. Suggesting that Chr08 (or a portion of it) is also sex linked. 

Hypotheses are as follows (in no particular order):

Error in mapping (seems unlikely given that this is transcriptome data)
Error in Xenopus assembly - also seems unlikely given quality of assembly
A real translocation somewhere between Xenopus and R. temporaria 
A fusion between Chr01 and Chr08 in this population. 

So we have 2 families from the same population Kilpisjarvi. They both in theory should have this pattern on Chr08.

So we want to know if there are fusions or translocations.

## Strategy:

NB: should first/in parrallel try to find sex of offspring - bioinformatically and/or with phenotypic sex assignments. Need this just to test that the Chr08 Fst signal is also found in these families (otherwise they are not very useful for this question). 

1. Linkage map the two families. Producing M and F maps for each.
2. Align mapped tags to R. temporaria genome - check for incongruence with LG assignment and assembly Chromosome assignments. 
3. Align tags (using temoraria scaffolds) to Xenopus and check again for incongruence with LG and assembly chromosome assignment.

Hopefully this information will allow us to test the hypotheses above. 


## Data

2 families from the same population:

- KAC1: 28 offspring, 2 parents
- KAC2: 35 offspring, 2 parents

1 lane of RADseq, using SpfI and MseI cutters. 

## Work

#### <b>Stacks</b>
Outside of this notebook, I ran <b>Stacks 1.48</b> on the two families, parameters.

- USTACKS: defaults
- CSTACKS: defaults
- SSTACKS: no params
- POPULATIONS: -p 2 (parents & offspring), -r 0.8, --vcf_haploypes, --write_single_snp, --min_maf 0.05, --max_obs_het 0.75. Populations was run separately for each family

<b>*Note</b> there was a problem when running populations, normal vcf output crashed. So I got around it by asking for a haplotype vcf and limiting to only one snp per tag. This may reduce the number of SNPs available for analyses, but likely not by much.

#### <b>Making input</b>
MSTmap input was made from the vcf haplotypes files for each family, using mendelian threshold of 0.75 and offspring presence threshold of 0.8. Number of tags per output below:

- KAC1 Male map: 1246
- KAC1 Female map: 1144
- KAC2 Male map: 1880
- KAC2 Female map: 1618
 
#### <b>Linkage mapping</b>

Linkage mapping was done using MSTmap (implemented in the R package ASmap v1.0-2) for each parent separately to give a total of 4 maps. For each map, multiple p-values were tested to find the optimal value. The "Optimum" value was taken to be that which gave the highest number of RADtags in linkage groups with relative sizes closest to that expected. i.e. 5 large linkage groups, 8 smaller ones, with a dropoff after the 13th linkage group to smaller blocks of unplaced markers. Optimal p-values were as follows:

- KAC1 Male map: 2e-4
- KAC1 Female map: 1e-4
- KAC2 Male map: 6e-5
- KAC2 Female map: 3e-6

I then outputted the names and positions of these mapped markers from R. 

Below I will:

1. Get sequences for each mapped tag (from the Stacks catalog)
2. Align sequences to the R.temporaria genome
3. Compare the sequence alignments with the linkage map orders

Step 3 should be the interesting part where any evidence for translocations or fusions between linkage groups should present themselves. .. . . . Lets do it!

The data and outputs for all analyses below can be found on Github [here](https://github.com/DanJeffries/Kilpisjarvi_Rtemp_linkage_maps) - the file directory structure will remain the same.

The RADtag IDs in the linkage map marker postion files come from the catalog. However, as in the linkage mapping proceedure (see [here](https://paperpile.com/shared/9iMAaD)), we duplicate markers in order to provdide both possible phases to MSTmap, some markers include the string "compli_" at the start. Thus I need to go through all markers, remove this if it exists, extract the tag ID, and get the sequence. I think I can do this all in one pass, if I first load the catalog into memory. 

So the script below goes as follows:

i. read all catalog tag sequences into a python dictionary - easy to query.  
ii. for each linkage mapped tag, search for the ID in the dictionary and write the sequence to the fasta output. 

```python
## Catalog file
import gzip

catalog = gzip.open("%s/MSTmap/batch_1.catalog.tags.tsv.gz" % wd, 'r').readlines()

## make catalog dictionary

catalog_dict = {}

for RADtag in catalog[1:]:
    RADtag_ID = RADtag.split()[2]
    RADtag_seq = RADtag.split()[8]
    catalog_dict[RADtag_ID] = RADtag_seq
```

Working with the KAC1 male map first

```python
## Now get IDs from the linkage map positions file
KAC1_M_positions = open("%s/MSTmap/KAC1/Male_map/KAC1_M_1e4_positions.txt" % wd, 'r' ).readlines()

## at the same time I will make a dictionary of the linkage mapping information for each marker too. 

KAC1_M_map_info_dict = {}

for marker in KAC1_M_positions[1:]:
    if marker.startswith("compli_"):
        marker_ID = marker.split()[0].split("_")[1]
    else:
        marker_ID = marker.split()[0].split("_")[0]
    
    marker_LG = marker.split()[1]
    marker_POS = marker.split()[2]
    
    KAC1_M_map_info_dict[marker_ID] = {}
    KAC1_M_map_info_dict[marker_ID]["LG"] = marker_LG
    KAC1_M_map_info_dict[marker_ID]["POS"] = marker_POS
        
## Now write the fasta for the mapped markers

mapped_marker_fasta = open("%s/MSTmap/KAC1/Male_map/KAC1_M_mapped_markers.fa" % wd, 'w')

counter = 0
for marker in KAC1_M_map_info_dict:
    mapped_marker_fasta.write(">%s\n%s\n" % (marker, catalog_dict[marker]))
    counter += 1
    
mapped_marker_fasta.close()

print "%s linkage mapped markers written to %s/MSTmap/KAC1/Male_map/KAC1_M_mapped_markers.fa" % (counter, wd)
```

Ok, so now I need to compare the alignment orders with the linkage map orders. I will define a function to do this, which requires the filtered blast hits in the dictionary object format outputted by MISC.BlastParseExtra(). First though, I will get the chromosome lengths to help with the plotting.

## 3. Plotting linkage maps vs physical marker alignments

```python
## First, get the lengths of all of the chromosomes in the assembly (for the plotting on the x axis)
genome_path = "/home/djeffrie/Data/Genomes/Rtemp_hybrid/Genome/Rtemp_anchored_all_scaffolds.fasta"

genome = SeqIO.parse(open(genome_path, 'r'),"fasta")

Chromosome_lengths = {}

for seq in genome:
    if "chr" in seq.id:
        Chromosome_lengths[seq.id] = len(seq.seq)
```

Here's the plotting function

```python
def Blast_vs_LM_plot(Blast_hits, LM_path, Chrom_assembly_lengths, Spacer):
    
    """
    Blast_hits:              The python dictionary from MISC.BlastParseExtra() containing filtered blast hits (DICT)
    LM_positions:            Full path to the text file containing the marker IDs (which must match those in <Blast_hits>) and their linkage map positions (STR)
    Chrom_assembly_lengths:  Python dictionary containing the chromosome lengths (Bp) for the assembly
    Spacer:                  The length of the space (Bp) between chromosomes on the x axis of the plot (INT)
    
    """
    
    from Bio import SeqIO
    from matplotlib import pyplot as plt

    ## Using the chromosome lengths provided I'm first making some prefix numbers to add to x axis coordinates for the plotting, the +1000 is just a spacer 
    
    Chrom_start = {}
    Chrom_start["chr1"] = 0
    Chrom_start["chr2"] = Chrom_assembly_lengths["chr1"] + Spacer
    Chrom_start["chr3"] = Chrom_start["chr2"] + Chrom_assembly_lengths["chr2"] + Spacer
    Chrom_start["chr4"] = Chrom_start["chr3"] + Chrom_assembly_lengths["chr3"] + Spacer
    Chrom_start["chr5"] = Chrom_start["chr4"] + Chrom_assembly_lengths["chr4"] + Spacer
    Chrom_start["chr6"] = Chrom_start["chr5"] + Chrom_assembly_lengths["chr5"] + Spacer
    Chrom_start["chr7"] = Chrom_start["chr6"] + Chrom_assembly_lengths["chr6"] + Spacer
    Chrom_start["chr8"] = Chrom_start["chr7"] + Chrom_assembly_lengths["chr7"] + Spacer
    Chrom_start["chr9"] = Chrom_start["chr8"] + Chrom_assembly_lengths["chr8"] + Spacer
    Chrom_start["chr10"] = Chrom_start["chr9"] + Chrom_assembly_lengths["chr9"] + Spacer
    Chrom_start["chr11"] = Chrom_start["chr10"] + Chrom_assembly_lengths["chr10"] + Spacer
    Chrom_start["chr12"] = Chrom_start["chr11"] + Chrom_assembly_lengths["chr11"] + Spacer
    Chrom_start["chr13"] = Chrom_start["chr12"] + Chrom_assembly_lengths["chr12"] + Spacer
    
    ## Now making a dictionary of colours for the plot using colormaps

    LM_positions = open(LM_path, 'r' ).readlines()

    ## First find number of LGs in the file
    
    LGs = []
    for line in LM_positions[1:]:
        marker_LG = line.split()[1]
        if marker_LG not in LGs:
            LGs.append(marker_LG)

    ## Then make a dictionary with the appropriate number of different colours for the plot
    
    LG_colours = {}
    num = 0
    iteration = 1/len(LGs)
    
    for LG in LGs:
        num+=iteration
        LG_colours[LG] = cm.Set1(num)
        
    
    ## Then get the linkage map positions for each marker
    
    print("Getting linkage map positions")
    
    LM_positions = open(LM_path, 'r' ).readlines()

    LM_info_dict = {}

    for marker in LM_positions[1:]:
        if marker.startswith("compli_"):
            marker_ID = marker.split()[0].split("_")[1]
        else:
            marker_ID = marker.split()[0].split("_")[0]

        marker_LG = marker.split()[1]
        marker_POS = marker.split()[2]

        LM_info_dict[marker_ID] = {}
        LM_info_dict[marker_ID]["LG"] = marker_LG
        LM_info_dict[marker_ID]["POS"] = marker_POS
    
    
    ## Now we can plot
    
    print "Plotting"
    
    
    fig = plt.figure(figsize = (60,20))
    ax1 = plt.axes(frameon=False)
    ax1.get_xaxis().set_visible(False)

    for chromosome in ["chr1", "chr2", "chr3", "chr4", "chr5", "chr6", "chr7", "chr8", "chr9", "chr10", "chr11", "chr12", "chr13"]:

        
        plt.plot([Chrom_start[chromosome],Chrom_start[chromosome] + Chrom_assembly_lengths[chromosome]], [-10, -10], 'k-', lw=2)
        plt.text((Chrom_start[chromosome] + 0.5*(Chrom_assembly_lengths[chromosome])) , -40, chromosome, fontsize = 20 ,horizontalalignment='center')
        
        for alignment in Blast_hits:

            if Blast_hits[alignment]["Ref_hit_id"] == chromosome:

                Chr = Blast_hits[alignment]["Ref_hit_id"]
                Pos = Chrom_start[Chr]+ int(Blast_hits[alignment]["Hit_start_coord"])
                #Pos = Chrom_cumul_lengths[Chr]+ int(KAC1_M_Filtered_blast_hits[alignment]["Hit_start_coord"])
                if alignment in LM_info_dict:
                    LG = LM_info_dict[alignment]["LG"]
                    LG_pos = LM_info_dict[alignment]["POS"]
                    #print LG_pos
                    ax1.scatter(Pos, float(LG_pos),s = 60, c = LG_colours[LG])
                
                
    ## legend
    patchList = []
    for LG in LG_colours:
        data_key = mpatches.Patch(color=LG_colours[LG], label=LG)
        patchList.append(data_key)

    plt.legend(handles=patchList, prop={'size': 20})
    plt.ylabel("Position cM", fontsize = 20)

    plt.show()
```

So, its pretty obvious that the linkage mapping using single families in MSTmap is not working particularly well here. For example Chr01 is represetned by 3 linkage groups, and this is the case for several other chromosomes. Also the male maps do not seem to be ordered correctly - I would expect there to be a low-high cM order across the chromosomes visible at the tips. It seems that there are just not enough offspring to make well resolved maps here.  

What we can see, especially from the male maps is that loci which are in one linkage groups on the whole map to one chromosome. Meaning there is no evidence here of large translocations or fusions in males. That said, our resolution is low. 

So the only way I see to proceed is to combine the two families to make a single map (for each sex), thus maximising the number of meioses available to the algorithm. To do this I will use LepMap3.

### LepMap3 family-combined linkage maps

I made the below linkage maps using LepMap3. The workflow is outlined [here](https://www.evernote.com/shard/s175/nl/19472134/df2e017a-01ce-48c9-8b7f-75f7a086197c?title=LepMap3%20workflow).

Below I will repeat the process performed with the MSTmap maps, first arranging the linkage map info (MarkerID, LG, Pos) and then plotting agains the alignments. 

```python
## This function makes the output required by the plotting function (it is the same format as genetic mapper needs)

def Genetic_mapper_in_maka(LepMap_outs_path):
    
    """
    Genetic_mapper_in_maka   < full/path/to/Lepmap_file >
    
    Makes two input files for genetic_mapper.pl, one for the male and one for the female map
    
    """
    

    LEPmap_outs = open(LepMap_outs_path, 'r').readlines()
    
    male_out_path = "%s_MALE_genetic_mapper.dat" % LepMap_outs_path.rpartition(".")[0]
    female_out_path = "%s_FEMALE_genetic_mapper.dat" % LepMap_outs_path.rpartition(".")[0]

    male_genetic_mapper_input = open(male_out_path, 'w')
    male_genetic_mapper_input.write("ID\tLG\tPOS\tLOD\n")
    female_genetic_mapper_input = open(female_out_path, 'w')
    female_genetic_mapper_input.write("ID\tLG\tPOS\tLOD\n")

    for line in LEPmap_outs:
        if line.startswith("#***"):
            LG = line.split()[3]
            #print LG
        elif not line.startswith("#") and "duplicate" not in line:
            marker_ID = line.split()[0]
            male_pos = line.split()[1]
            female_pos = line.split()[2]

            male_genetic_mapper_input.write("%s\t%s\t%s\n" % (marker_ID,LG,male_pos))
            female_genetic_mapper_input.write("%s\t%s\t%s\n" % (marker_ID,LG,female_pos))

    male_genetic_mapper_input.close()
    female_genetic_mapper_input.close()
```

### (Joint) Female map

```python
### Get LepMap tag references and map info from the genetic mapper format file

linkage_map = open("%s/Lepmap/map_lod_10_inf_23_js9_ordered_FEMALE_genetic_mapper.dat" % wd, 'r').readlines()

Lepmap_map_info = {}
LepMap_refs = []

for line in linkage_map:
    if line.startswith("#") == False and "ID" not in line:
        LepMap_ID = line.split()[0]
        LepMap_refs.append(LepMap_ID)
        Lepmap_map_info[LepMap_ID] = {}
        Lepmap_map_info[LepMap_ID]["LG"] = line.split()[1]
        Lepmap_map_info[LepMap_ID]["POS"] = line.split()[2]
        
        
print len(LepMap_refs), "SNPs in the map"

### Now we get the order of the markers in the VCF and their true marker ID.

VCF = open("%s/Lepmap/batch_1.haplotypes.vcf" % wd, 'r').readlines()

marker_ID_dict = {}

marker_number = 1
for line in VCF:
    if not line.startswith("#"):
        VCF_marker_ID = line.split()[2]
        
        marker_ID_dict[str(marker_number)] = VCF_marker_ID
        
        marker_number += 1
        
print len(marker_ID_dict), "SNPs in the VCF"

### Next find the true reference of the markers that made it into the map and write a new map file with the true references

Mapped_true_SNP_IDs = []

true_ID_map_info = open("%s/Lepmap/map_lod_10_inf_23_js9_ordered_FEMALE_genetic_mapper_true_IDs.dat" % wd, 'w')
true_ID_map_info.write("ID\tLG\tPOS\n")

for LepMap_ref in LepMap_refs:
    if LepMap_ref in marker_ID_dict:
        True_ID = marker_ID_dict[LepMap_ref]
        Mapped_true_SNP_IDs.append(True_ID)
        true_ID_map_info.write("%s\t%s\t%s\n" % (True_ID, Lepmap_map_info[LepMap_ref]["LG"], Lepmap_map_info[LepMap_ref]["POS"]))

print len(Mapped_true_SNP_IDs), "true SNP IDs recovered"
print len(set(Mapped_true_SNP_IDs)), "RAD tags contain these SNPs"

true_ID_map_info.close()

## Now we extract the sequences of those tags from the catalog file

catalog_path = "%s/MSTmap/batch_1.catalog.tags.tsv.gz" % wd

MISC.fasta_maka(set(Mapped_true_SNP_IDs), catalog_path)
```
