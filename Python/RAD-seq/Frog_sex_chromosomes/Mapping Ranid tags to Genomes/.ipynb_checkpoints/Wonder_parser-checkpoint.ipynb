### This is a parser for the output of the Wonder Pipeline v.07

Create a function to parse the results file and return a dictionary containing all information that can be accessed by future functions

```python
from pprint import pprint as pp
```

```python
results_file = open("/home/djeffrie/Data/Mapping_SL_tags_to_genomes/Sex_linked_tags/R_arvalis/results-arvalis_dan.txt", 'r').readlines() ## not going to be massive files so can read in all at once. 

Master = {} ## the master dictionary for all output info. 

for line in results_file[1:10]:
    record_name = line.split()[0]
    stage = line.split()[1]
    
    ## Make first level structure 
    
    if record_name not in Master:
        Master[record_name] = {}
    
    Master[record_name]["Blastn_Genome_1"] = {}
    Master[record_name]["Blastn_Genome_2"] = {}
    Master[record_name]["Blastn_Genome_3"] = {}
        
    Master[record_name]["Proteome_1"] = {}
    Master[record_name]["Proteome_2"] = {}
    Master[record_name]["Proteome_3"] = {}
    
    
    if stage == "200":
        
        ## Get blast results to the first genome
        
        blast_match_id = line.split()[4] ## ID for the blast hit for this record
        
        Master[record_name]["Blastn_Genome_1"][blast_match_id] = {}
        Master[record_name]["Blastn_Genome_1"][blast_match_id]["subject_id"] = line.split()[5]
        Master[record_name]["Blastn_Genome_1"][blast_match_id]["perc_id"] = line.split()[6]
        Master[record_name]["Blastn_Genome_1"][blast_match_id]["length"] = line.split()[7]
        Master[record_name]["Blastn_Genome_1"][blast_match_id]["Eval"] = line.split()[14]
        Master[record_name]["Blastn_Genome_1"][blast_match_id]["Magnitude"] = line.split()[18]
           

    
    pp(Master)




```
