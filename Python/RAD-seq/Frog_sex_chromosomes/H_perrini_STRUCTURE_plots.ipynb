## K2 iteration 2

```python
str_outs = open("/home/djeffrie/Data/RADseq/H_perrinii/Populations_kept_largepops_p12_r0.8_m6/structure/structure_cline/Structure_OUTS_harvester/batch_1.structure_OUTS_K2_1.tsv_f", 'r').readlines()

fig = plt.figure(figsize = (40,10))

## get sample names in right order
samples = open("/home/djeffrie/Data/RADseq/H_perrinii/Populations_kept_largepops_p12_r0.8_m6/samples_sorted_by_lat.txt", 'r').readlines()
sample_names = []

for line in samples:
    sample_names.append(line.split()[0])


## get the assignment information into a dictionary
  
assign_dict = {}
    
for line in str_outs:
    if line.startswith("#"):
        if "Label" not in line:
            assign_dict[line.split()[2]] = (line.split()[6], line.split()[7])
            
x_coord = 0

for sample in sample_names:
    plt.bar(x_coord, float(assign_dict[sample][0]), color = (0.12, 0.31, 0.57))
    plt.bar(x_coord, float(assign_dict[sample][1]), bottom = float(assign_dict[sample][0]), color = (0.182,0.78,0.037))
            
    x_coord += 1
           
            
plt.xticks(range(x_coord), sample_names, rotation = -90)
plt.savefig("/home/djeffrie/Data/RADseq/H_perrinii/Populations_kept_largepops_p12_r0.8_m6/structure/K2.pdf")
plt.show()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt


N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, color='#d62728', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```
