## Finding gene capture targets from <i>Rana temporaria</i> transcripts annotated using their <i>Xenopus tropicalis</i> 1 to 1 orthologs.

Here I want to find some more target sequences for our capture project, by using a previously made transcriptome for R. temporaria which was annotated (By Melissa Toups) against the Xenopus genome. 

I am interested in 3 chromosomes, Chr01, Chr03 and Chr05. I would like to sample genes along the length of these chromosomes. This would hopefully allow us to find sex linked signals and better characterise the differentiation among the X and Y on these chromosomes in the focal species. I could also use the sequences to assess dn/ds, which would be something we have not yet done. 


```python
from __future__ import division
from Bio import SeqIO
import numpy as np
```

So I want to first plot how many genes I have per chromosome of interest. 

```python
one_to_ones = SeqIO.parse(open("/home/djeffrie/Data/Sexome/Annotated_rtemp_transcriptome/TFDtrim_OrthOnly_wChrom.fasta.transdecoder.LI2.cds", 'r'), "fasta")

Chrom_dict = {}
Chrom_dict["01"] = []
Chrom_dict["03"] = []
Chrom_dict["05"] = []

for orth in one_to_ones:

    gene_name = orth.id.split("_")[5]
    xen_chrom = orth.id.split("_")[6]
    xen_coord = orth.id.split("_")[7]
    
    if xen_chrom == "Chr01":
        Chrom_dict["01"].append(int(xen_coord))
    elif xen_chrom == "Chr03":
        Chrom_dict["03"].append(int(xen_coord))
    elif xen_chrom == "Chr05":
        Chrom_dict["05"].append(int(xen_coord))
    
```

Ok, so we have plenty of genes to chose from, spread across the entirety of the linkage groups. So now I just need a good way of finding a gene for each of a given number of windows along the genome.  How best to do this? 

### Sampling strategy

My strategy will be to first cut the length of the chromosome into sections. The number of sections can be increased or decreased depending on how many genes we can afford to sample. 

Next I will assign each gene to its relevent section. 

And finally, I will choose the gene closest to the centre of each section (to avoid having genes at the end of one section and the beginning of the next, i.e. very clumped).

```python
## Split the chromosome length into sections

N_sections = 100

length_of_chrom = max(Chrom_dict["01"])
window_size = np.round(length_of_chrom/N_sections, 0)

Section_dict = {}

prev_section_stop = 0
for section_index in range(1,N_sections):
    Section_dict[str(section_index)] = (prev_section_stop+1, prev_section_stop+(window_size/2), prev_section_stop+window_size)
    prev_section_stop = prev_section_stop+window_size

    
## Now assign genes to the sections

one_to_ones = SeqIO.parse(open("/home/djeffrie/Data/Sexome/Annotated_rtemp_transcriptome/TFDtrim_OrthOnly_wChrom.fasta.transdecoder.LI2.cds", 'r'), "fasta")

gene_sections = {}

for i in range(1,N_sections):
    gene_sections[str(i)] = {}

for orth in one_to_ones:
    
    gene_name = orth.id.split("_")[5]
    xen_chrom = orth.id.split("_")[6]
    xen_coord = orth.id.split("_")[7]
    
    if xen_chrom == "Chr01":
        for section in Section_dict:
            if Section_dict[section][0] <= int(xen_coord) <= Section_dict[section][1]:
                
                gene_sections[str(section)][int(xen_coord)] = gene_name
                
## Note I have structured the dictionary this way so it will work with the lambda function below. 

target_genes = {}

for section in range(1,N_sections):
    section = str(section)
    if len(gene_sections[section]) > 0:

        midpoint = Section_dict[section][1]
        closest = (min(gene_sections[section], key=lambda x:abs(x-int(midpoint))))

        #print section, int(midpoint), closest, gene_sections[section][closest]
        
        target_genes[gene_sections[section][closest]] = closest
        
    #else:
        #print section, "empty"
    
## Now I want to check the distribution of those genes along the chromosome

plt.figure(figsize = (30,5))

plt.bar(target_genes.values(), [1]*len(target_genes.values()), width = 150000)
plt.title("Chromosome 1")
plt.xlabel("Base position")
plt.ylim(0,1.25)
plt.show()

## Finally, output these selected genes to a fasta for each chromosome

one_to_ones = SeqIO.parse(open("/home/djeffrie/Data/Sexome/Annotated_rtemp_transcriptome/TFDtrim_OrthOnly_wChrom.fasta.transdecoder.LI2.cds", 'r'), "fasta")
target_gene_fasta = open("/home/djeffrie/Data/Sexome/transcriptome_targets/Chr01_transcriptome_targets_%s.fasta" % N_sections, 'w') 

for orth in one_to_ones:
    
    gene_name = orth.id.split("_")[5]
    
    if gene_name in target_genes:
        SeqIO.write(orth, target_gene_fasta, "fasta")
        
target_gene_fasta.close()
print "\nTarget gene CDS sequence written to file"
```
