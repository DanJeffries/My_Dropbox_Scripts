```python

for target_LG in LGs:
    
    linkage_map = open("/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/MAP23_ordered1/Map_23_ordered_1.txt", 'r').readlines()

    LG_positions = []
    
    for i in linkage_map:

        if i.startswith("#***"):

            LG = i.split()[3]

        if LG == target_LG and not i.startswith("#"):

            pos = i.split()[1]
            LG_positions.append(float(pos))
            
    plt.figure(figsize = (15, 5))        
    
    plot = sns.kdeplot(np.array(LG_positions), color = "black", shade = True, bw = 0.01)
    
    plt.xlim(0,101)
    plt.ylim(0,2.5)
    plt.title(target_LG)
    
    plt.savefig("/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/MAP23_ordered1/LG_%s_Male_SNP_density.svg" % target_LG)
    
    
    
    
```
