## Here I will parse the output of the LepMap3 module IBD - which calculates the relatedness between individuals in a family to check that they are indeed full or half sibs and parents. 

```python
% matplotlib inline
from matplotlib import pyplot as plt
```

```python
IBD_outs = open("/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/IBD_outs.txt", 'r').readlines()
```

## ST01

```python
fam = "ST01"
mother = "ST01_F_all"
father = "ST01_M_all"

IBDs = []
labs = []

for i in IBD_outs:
    if i.split()[0] == mother :
        if fam in i.split()[1] and i.split()[1] != father:
            IBDs.append(float(i.split()[2]))
            labs.append(i.split()[1])
    
    elif i.split()[0] == father:
        if fam in i.split()[1] and i.split()[1] != mother:
            IBDs.append(float(i.split()[2]))
            labs.append(i.split()[1])
    
    elif i.split()[1] == mother :
        if fam in i.split()[0] and i.split()[0] != father:
            IBDs.append(float(i.split()[2]))
            labs.append(i.split()[0])
    
    elif i.split()[1] == father :
        if fam in i.split()[0] and i.split()[0] != mother:
            IBDs.append(float(i.split()[2]))
            labs.append(i.split()[0])
        
            
plt.figure(figsize = (30,10))
plt.bar(range(1,len(IBDs)+1,1),IBDs, align = "center")
plt.xticks(range(1,len(IBDs)+1,1),labs, rotation = 90)
plt.show()
```
