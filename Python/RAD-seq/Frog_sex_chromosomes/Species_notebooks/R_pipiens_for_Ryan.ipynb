## Sample set:

We generated ddRADseq data in Lausanne which were processed with the program Stacks before the analyses herein. Stacks essentially just constructs loci from the raw sequence reads, and calls Single nucleotide polymorphisms (SNPs) between samples. These SNP markers are what is used in the analyses below. At this stage, they are anonymous, which is to say that their location in the genome is not known.

There are 12 populations in this data, generally with 8-10 samples per population. The samples come from and are described in [Oâ€™Donnell & Mock (2012)](https://link.springer.com/article/10.1007/s10592-012-0384-4).

# Subsetting the data

Popualtion structure often confounds the search for sex linked markers for two main reasons. 

1) Population structure can produce false positives  
2) Sex chromosome haplotypes can vary between populations

For that reason, often the only way to find sex linked markers is to search for them in groups of closely related individuals. Unfortunately, this has to be balanced with the reduction in power that comes with using fewer males and females. So below, I try to group populations that are closely related to each other (based on the above PCA) in order to maximise the number of males and females in each analysis.

# SUMMARY

This is a somewhat frustrating sample set. Of course in hindsight its easy to see why. When analysing data across several populations, the only sex linked markers that can be found are those that are sex linked in all of them, which means that they were present in the sex chromosome of the ancestor of all populations. For species like those in Ranidae, with very homomorphic sex chromosomes, there are likely to be very few ancestrally sex linked markers, and furthermore, through mechanisms like the fountain of youth, sex chromosome haplotypes can become "refreshed" (through recombination between X and Y in XY females if they occur). This happens in a lineage specific way so that, after some time, the sex chromosome haplotypes among popualtions no longer share any similarity. 

So to go further with this species, RADseq data from more individuals per population would be required. I would say between 10-15 males and females per population. Even more if populations are particularly outcrossed. Alternatively, sexed families would be the ideal option. 

The RADdata herein however, is certainly viable for phylogeographic analyses, judging just by the PCA, the resolution is of course much higher than with previous markers, and seems to agree with previous findings.
