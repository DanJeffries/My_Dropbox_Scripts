### Here I will use the RADtags in the LepMap linkage map (made by combining 6 temporaria families) to find homology between R. temporaria scaffolds and Xenopus reference chromosomes

Do do this we:

1. Map the tags in the linkage map to the current Rana temporaria genome. 
2. For scaffolds that have hits, map them to the xenopus genome
3. Trace back through the mappings to find which LG belongs to which Xenopus chromosome.

```python
%matplotlib inline
from __future__ import division
import sys
sys.path.append('/home/djeffrie/Dropbox/My_Dropbox_Scripts/Python/My_Modules/')
import MISC_RAD_tools as MISC
from matplotlib import pyplot as plt
from matplotlib import cm
from Bio import SeqIO
```

## Step 1.

    1.a) LepMap doesn't use the real tag IDs in the map, it gives them its own reference from 1 - N according to their order in the vcf. So the first step is to find which of the tags in the VCF made it into the map, and then extract those tags from the catalog.

```python
### Get LepMap tag references from the map file

linkage_map = open("/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/Map_23_ordered_1.txt", 'r').readlines()

LepMap_refs = []

for line in linkage_map:
    if not line.startswith("#"):
        LepMap_refs.append(line.split()[0])
        
print len(LepMap_refs), "SNPs in the map"
```

```python
### Now we get the order of the markers in the VCF and their true marker ID.

VCF = open("/home/djeffrie/Data/RADseq//R_temp_fams/Populations_all_kept/batch_1.vcf", 'r').readlines()

marker_ID_dict = {}

marker_number = 1
for line in VCF:
    if not line.startswith("#"):
        marker_ID = line.split()[2]
        
        marker_ID_dict[str(marker_number)] = marker_ID
        
        marker_number += 1
        
print len(marker_ID_dict), "SNPs in the VCF"
```

```python
### Next find the true reference of the markers that made it into the map

Mapped_true_SNP_IDs = []

for Map_snp in LepMap_refs:
    if Map_snp in marker_ID_dict:
        Mapped_true_SNP_IDs.append(marker_ID_dict[Map_snp])

print len(Mapped_true_SNP_IDs), "true SNP IDs recovered"
print len(set(Mapped_true_SNP_IDs)), "RAD tags contain these SNPs"
```

```python
## Now we extract the sequences of those tags from the catalog file

catalog_path = "/home/djeffrie/Data/RADseq/R_temp_fams/batch_1.catalog.tags.tsv.gz"

MISC.fasta_maka(set(Mapped_true_SNP_IDs), catalog_path)
```

### These tags were then mapped to the Rtemp genome using blastn the xml output was retrieved for use below.

We now filter the mappings to Rtemp, and keep only the confident ones based on e-value. For these mappings the "MISC.BlastParseExtra" function will output a segment of that R. temporaria scaffold. Here I will use a segment size of ~4kb (2kb either side of the mapping). However in some cases the scaffold will end before this 2kb length is reached, thus the segment will often be shortened.

```python
Tags_to_Rtemp_blastn_outs = "/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/LM_tags_to_Rtemp.xml"
Rtemp_fasta = "/home/djeffrie/Data/Genomes/Rtemp/Rtemp_gapfilled.iter2.fa"
```

```python
## Filtering the blastn mappings using minimum e-val threshold of 1e-20

best_hit_crit = 1e-5 ## the top hit for a tag must be 5 orders of magnitute better than the next best hit
Eval_threshold = 1e-20 ## the minimum evalue for a hit
Window = 2000 ## the size (each side of the mapping) of the fragment to retreive from the fasta for the next mapping stage. 

RADtags_to_Rtemp_Filtered_blasts = MISC.BlastParseExtra(Tags_to_Rtemp_blastn_outs, Rtemp_fasta , best_hit_crit, Eval_threshold, 1 ,Window)
```

### Next mapped the retained 4kb scaffolds from the eval = 1e-20 mapping to the Nanorana genome. 

```python
Rtemp_to_Nano_blastn_outs = "/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/Rtemp_to_Nano_blastn.xml"
Nano_fasta = "/home/djeffrie/Data/Genomes/Nano/Nanorana_parkeri.genome.v2.fa"

## Filtering the blastn mappings using minimum e-val threshold of 1e-20

best_hit_crit = 1e-5 ## the top hit for a tag must be 5 orders of magnitute better than the next best hit
Eval_threshold = 1e-20 ## the minimum evalue for a hit
Window = 4000 ## the size (each side of the mapping) of the fragment to retreive from the fasta for the next mapping stage. 

Rtemp_nano_filtered_blasts = MISC.BlastParseExtra(Rtemp_to_Nano_blastn_outs, Nano_fasta , best_hit_crit, Eval_threshold, 1 ,Window)
```

So using a higher mapping stringency gets rid of about 300 mapped sequences. Not a big difference, but migh help false positives. Can come back to this later if false positives are a big issue.

### Next mapped the retained 4kb scaffolds from Nanorana to the Xenopus genome. 

```python
Nano_Scaffs_to_xen_blastn_outs = "/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/Nano_to_Xen_blastn.xml"

Xen_fasta = "/home/djeffrie/Data/Genomes/Xenopus_genome/Xtropicalis.v9.repeatMasked.fa"


best_hit_crit = 1e-5 ## the top hit for a tag must be 5 orders of magnitute better than the next best hit
Eval_threshold = 1e-20 ## the minimum evalue for a hit
Window = 2000 ## the size (each side of the mapping) of the fragment to retreive from the fasta for the next mapping stage. 

Rtemp_to_Xen_Filtered_blasts = MISC.BlastParseExtra(Nano_Scaffs_to_xen_blastn_outs , Xen_fasta , best_hit_crit, Eval_threshold, 1 ,Window)
```

### Now find which RADtags map to which Xenopus chromosome.

```python
## Get the location of the scaffold in the Xenopus genome and the position of the marker that maps to it in the R. temporaria linkage map.

Blast_dictionary = {}

## Step one - find which Rtemp scaffs each RADtag hit

for RADtag in RADtags_to_Rtemp_Filtered_blasts:
    
    if RADtag not in Blast_dictionary:
        Blast_dictionary[RADtag] = {}
        Blast_dictionary[RADtag][RADtags_to_Rtemp_Filtered_blasts[RADtag]["Ref_hit_id"]] = {}
        
## Step two - find which Rtemp_scaffs hit which Nanorana scaffs

for RADtag in Blast_dictionary:
    for Rtemp_scaff in Blast_dictionary[RADtag]:
        if Rtemp_scaff in Rtemp_nano_filtered_blasts:
            Blast_dictionary[RADtag][Rtemp_scaff][Rtemp_nano_filtered_blasts[Rtemp_scaff]["Ref_hit_id"]] = None
            
## Step three - find which Nanorana scaffs hit Xenopus

for RADtag in Blast_dictionary:
    for Rtemp_scaff in Blast_dictionary[RADtag]:
        for Nano_scaff in Blast_dictionary[RADtag][Rtemp_scaff]:
            if Nano_scaff in Rtemp_to_Xen_Filtered_blasts:
                Xen_chrom = Rtemp_to_Xen_Filtered_blasts[Nano_scaff]["Ref_hit_id"]
                Xen_srt_pos = Rtemp_to_Xen_Filtered_blasts[Nano_scaff]["Hit_start_coord"]
                Xen_end_pos = Rtemp_to_Xen_Filtered_blasts[Nano_scaff]["Hit_end_coord"]
                Blast_dictionary[RADtag][Rtemp_scaff][Nano_scaff] = (Xen_chrom, Xen_srt_pos, Xen_end_pos)
```

### Now look at which tags on each linkage group map to which Rtemp scaffold, assign them to Xenopus chromosomes and output these scaffolds with this information in their headers.

```python
LM_to_Xen_mapping_summary = open("/home/djeffrie/Data/RADseq/R_temp_fams/Populations_all_kept/Lepmap/LepMap3/MAP_23/LepMap_Map_23_to_Xen_mapping_summary_male.txt", 'w')

for LG in LG_to_Chrom_dict:
    for true_tag in TRUE_RADtag_LM_info_dict:
        
        Rtemp_scaff = "-"
        Nano_scaff = "-"
        Xen_chrom = "-"
        Xen_strt = "-"
        Xen_end = "-"
        tag_lg_pos = "-"
                
        if TRUE_RADtag_LM_info_dict[true_tag][0] == LG:
            tag_lg_pos = TRUE_RADtag_LM_info_dict[true_tag][1]
            if true_tag in Blast_dictionary: ## it hits Rtemp
                for Rtemp_scaff in Blast_dictionary[true_tag]:
                    if len(Blast_dictionary[true_tag][Rtemp_scaff]) == 1: # if it hits Nano
                        Nano_scaff = Blast_dictionary[true_tag][Rtemp_scaff].keys()[0]
                        if Blast_dictionary[true_tag][Rtemp_scaff][Nano_scaff] != None:
                            Xen_chrom = Blast_dictionary[true_tag][Rtemp_scaff][Nano_scaff][0]
                            Xen_strt = Blast_dictionary[true_tag][Rtemp_scaff][Nano_scaff][1]
                            Xen_end = Blast_dictionary[true_tag][Rtemp_scaff][Nano_scaff][2]
                        
                    else:
                        Nano_scaff = "-"
                        Xen_chrom = "%s_inf" % LG_to_Chrom_dict[LG]
                        Xen_strt = "-"
                        Xen_end = "-"
                        
                    LM_to_Xen_mapping_summary.write( "%s\t%s\t%s\t%s\t%s\t%s\t%s\t%s\n" % (true_tag, Rtemp_scaff, Nano_scaff, Xen_chrom, Xen_strt, Xen_end, LG, tag_lg_pos))
LM_to_Xen_mapping_summary.close()
```
