###1. Ustacks tests - Using Incremental_U

Ustacks is the first module in the de-novo pipeline of Stacks. Its function is to cluster reads, based on their sequence similarity, into alleles and loci. This process is highly parameterised. So far, Incremental_U supports tests of three of the major parameters:
    

    M: Maximum distance (in nucleotides) allowed between stacks (Stacks default 2).
    m: Minimum depth of coverage required to create a stack (Stacks default 3).
    MS: (--max_locus_stacks) maximum number of stacks at a single de novo locus (Stacks default 3).


```python
### Fastq files are here:
Incrememntal_test_directory = "/home/djeffrie/Data/Incremental/raw_data/"  ## Note - stupidly fails if not missing the last "/" from path! Need to fix that . . . 
```

```python
## INPUT INCREMENTAL PARAMETERS HERE --( Can be in a parameter file for Cline version )----------------------

parameters = ["M", "m", "MS"]                                   ## List, eg ["M", "m", "MS"]. If only testing 1 param, this must still be a list e.g. ["M"]
M_vals =   [2,3,4,5,6]                                          ## List, eg [1,2,3,4]
m_vals =   [2,3,4,5,6]                                          ## List, eg [1,2,3,4]
MS_vals =  [2,3,4]                                              ## List, eg [1,2,3,4]
threads =  7                                                    ## Int. eg 7
parent_dir = Incrememntal_test_directory                        ## Absolute Dir containing the demultiplexed raw read fasta/fastq files
run_cline_switch =  1                                           ## to switch on (1) to run stacks cline or off (0) to run pipeline on existing Stacks outputs


## Making input dictionary --------------------------------------

Setup_dict = {}
Setup_dict["parameters"] = {}
Setup_dict["threads"] = threads
Setup_dict["parent_directory"] = parent_dir
Setup_dict["run_cline_switch"] = run_cline_switch
Setup_dict["verbose"] = False

for param in parameters:
    if "M" in param:
        Setup_dict["parameters"][param] = M_vals
    if "m" in param:
        Setup_dict["parameters"][param] = m_vals
    if "MS" in param:
        Setup_dict["parameters"][param] = MS_vals

#pp = pprint.PrettyPrinter(indent=2)
#pp.pprint(Setup_dict)
```

```python
Incremental_U_outputs = IncU.IncreMental_U(Setup_dict)
```

### There is now an intermediate step before we can proceed to the Cstacks tests:

Cstacks requires Ustacks outputs as its inputs. Obviously which Ustacks outputs we use will affect the Cstacks results. This is perhaps not a really big deal as the inputs for Cstacks will always be the same for all Cstacks tests - so the patterns across the Cstacks parameter values should be roughly the same regardless of the Ustacks inputs used. However, it would be good to allow the user to control this if they want. 

For running the pipeline module by module separately, it is easy, we include a function (which I think I already have) to run Ustacks a final time for whatever parameter values the user likes.

But if we want to add an automated wrapper for the whole Incremental pipeline, we need to deal with this another way. 

I think we could design the pipeline to run in two ways:

    1. Automatic - The user defines all test parameter values and all default values for each module (Ustacks, Cstacks and Populations) in a single parameter file. Then, at the end of each step the final run of that model is run using the default parameter values or a set of specified values specifically for the final run.
    2. Modular - The user runs the pipeline one module at a time, and manually runs the final commands for each module using a function that we include.

## We need to make another decision here. . . . 

The Cstacks step is quite long - the more samples you use, the more loci are added to the catalog. The more loci in the catalog, the more time it takes to search further samples against that catalog. Thus computation time of this step can increase exponentially as more samples are added. Further more, it only possible to construct venn diagrams for 4 subsets of data - any higher and it becomes geometrically impossible to make a 2D plot of all overlaps. So I find that it is best at this stage to choose subsets of the dataset to test. 

For example, as in the test data - the dataset may contain different species and hybrids. I would therefore run Incremental_C several times for different combinations of 3 or 4 samples at a time, comparing conspecifics in one run and comparing individuals from separate species, and potentially their hybrids in anoter run. This keeps computation time down, and informativeness up. 

So in the automatic version of Incremental - I propose that the default is to randomly choose 3 sets of test samples, each with a different combination of individuals in the test data. We can however, allow the user to specify combinations of samples to test in this step in the parameter file. 

Thoughts welcome here! 

As an interesting test case, the 3 sample comparison above uses one pure C. carassius sample, one pure C. auratus sample and one F1 hybrid. You can see in the Venn diagrams how the hybrid contains more loci than either of the other two, and that many loci are shared just by the hybrid, and one of the parental species, but not the other. This is indicative of null alleles, where a Restriction site has been lost (or gained) in one of the parental species. The hybrid therefore inherits one allele containing a RAD site and one allele lacking a RAD site, meaning that it is essentially hemizygous at this locus. Although you can't see it in this step, these loci are likely to look almost entirely homozygous in the hybrid, and of course only half the coverage of the normal loci, as only one allele is being sequenced. This is a good example of how these tests can be informative for complex datasets. 

I will probably try to include something like this in the paper (I am thinking somewhere like Mol.Ecol. resources would be good).

## Populations tests - using Incremental_Pop

The Populations module of Stacks does a lot. It takes all of the outputs from the previous three Stacks modules, collects genotype and haplotype information for a set of specified samples and can output that information into a number of different formats. It also performs lots of population genetics analyses.

So far, Incremental deals with 3 parameters:

    p: minimum number of populations a locus must be present in to process a locus.
    r: minimum percentage of individuals in a population required to process a locus for that population.
    m: specify a minimum stack depth required for individuals at a locus.
